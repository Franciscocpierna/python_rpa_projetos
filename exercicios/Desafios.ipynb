```python
"""
1 - Faça um Programa que verifique se uma letra digitada é uma vogal ou consoante.

O programa deverá seguir as seguintes regras:

- Solicitar ao usuário que insira uma letra.
- Verificar se o caractere inserido é uma letra do alfabeto.
- Caso o usuário tenha inserido um caractere válido (ou seja, uma 
            letra do alfabeto), o programa deverá identificar e informar 
            se essa letra é uma vogal ou uma consoante.
- Caso o usuário insira um valor inválido (como números, símbolos ou 
            mais de um caractere), o programa deverá exibir uma mensagem de 
            erro, informando que a entrada é inválida e pedindo para que 
            ele insira apenas uma letra.
"""

# Define uma função 'verificar_letra' que recebe 'letra' 
        # como parâmetro.
# O propósito desta função é determinar se uma entrada 
        # única é uma vogal ou consoante,
        # e garantir que a entrada seja um caractere alfabético válido.
def verificar_letra(letra):
    
    # Cria uma lista contendo todas as vogais do alfabeto, 
            # tanto em caixa alta quanto baixa.
    # Isso permite que a função verifique de maneira eficaz e 
            # sem distinção de maiúsculas ou minúsculas.
    #vogais = ['a', 'e', 'i', 'o', 'u', 'A', 'E', 'I', 'O', 'U']
    vogais="aeiouAEIOU"
    # Primeiro, verifica se a entrada é um caractere alfabético e 
            # se consiste em apenas um caractere.
    # Utiliza 'letra.isalpha()' para garantir que a entrada 
            # seja alfabética, evitando números ou símbolos.
    # Utiliza 'len(letra) == 1' para assegurar que a entrada 
            # seja apenas um único caractere,
            # o que é essencial para a validação de vogal ou consoante.
    if letra.isalpha() and len(letra) == 1:
        
        # Se a letra está contida na lista de vogais definida, 
                # então é identificada como vogal.
        # A comparação 'letra in vogais' efetua uma busca na 
                # lista 'vogais' para verificar se a 'letra'
                # corresponde a qualquer elemento na lista.
        if letra in vogais:
            return f"A letra '{letra}' é uma vogal."
            
        # Caso a letra não esteja na lista de vogais, por 
                # eliminação, é uma consoante.
        # Isso ocorre apenas se a letra é alfabética e não é 
                # uma das vogais listadas anteriormente.
        else:
            return f"A letra '{letra}' é uma consoante."
            
    # Se a entrada não satisfizer as condições de ser 
            # alfabética e de ter apenas um caractere,
    # a função retorna uma mensagem de erro, pedindo 
            # uma entrada válida.
    else:
        return "Entrada inválida! Por favor, insira apenas uma letra."

# Solicita ao usuário que insira uma letra do 
        # alfabeto para verificação.
# Esta etapa interage diretamente com o usuário, 
        # pedindo uma entrada simples de texto.
letra_digitada = input("Digite uma letra: ")

# A função 'verificar_letra' é chamada com a entrada do 
        # usuário, e o resultado é armazenado em 'resultado'.
# Este resultado determinará se a entrada é uma vogal, 
        # consoante ou se foi inválida.
resultado = verificar_letra(letra_digitada)

# O resultado da verificação é impresso, mostrando ao 
        # usuário se a letra inserida é uma vogal, consoante
        # ou se a entrada foi inválida.
print(resultado)
```

```python
"""
2 - Faça um Programa que leia um número e exiba o dia correspondente da semana.

O programa deverá seguir as seguintes regras:

Solicitar ao usuário que insira um número de 1 a 7.

O programa deve exibir o dia da semana correspondente ao número 
        inserido, conforme a tabela abaixo:

1 - Domingo
2 - Segunda-feira
3 - Terça-feira
4 - Quarta-feira
5 - Quinta-feira
6 - Sexta-feira
7 - Sábado

Caso o usuário insira um número fora desse intervalo (ou seja, um 
        número menor que 1 ou maior que 7), o programa deverá exibir a 
        mensagem "Valor inválido", indicando que o número não 
        corresponde a um dia da semana.
        
O programa deve garantir que a entrada seja um número válido. 

Se o usuário inserir um valor que não seja numérico (como letras 
        ou símbolos), o programa deve exibir uma mensagem de erro apropriada.
"""

# Define a função 'dia_da_semana' para determinar o dia da 
        # semana com base em um número.
def dia_da_semana(numero):
    
    # Inicializa um dicionário 'dias' onde as chaves são 
            # números inteiros (1 a 7),
    # e os valores são strings representando os dias da 
            # semana correspondentes a cada número.
    # O uso de um dicionário permite uma recuperação 
            # eficiente e direta do dia da semana
            # associado a cada número, facilitando a manutenção e a 
            # escalabilidade do código.
    dias = {
        1: "Domingo",
        2: "Segunda-feira",
        3: "Terça-feira",
        4: "Quarta-feira",
        5: "Quinta-feira",
        6: "Sexta-feira",
        7: "Sábado"
    }
    
    # Verifica se o número fornecido está dentro do 
            # intervalo definido (1 a 7).
    # A checagem é feita verificando a presença do número 
            # nas chaves do dicionário 'dias'.
    # Essa verificação garante que apenas números válidos 
            # sejam processados,
            # evitando erros ou comportamentos indesejados.
    if numero in dias:
        
        # Se o número é válido, a função retorna uma string 
                # formatada indicando o dia da semana correspondente 
                # ao número, utilizando o valor recuperado do dicionário 'dias'.
        return f"O dia correspondente ao número {numero} é {dias[numero]}."
        
    else:
        
        # Caso o número não esteja no intervalo válido (1 a 7), 
                # retorna uma mensagem de erro,
                # solicitando ao usuário que forneça um número 
                # dentro do intervalo correto.
        return "Valor inválido! Por favor, insira um número entre 1 e 7."

# Bloco principal que executa a funcionalidade do programa.
try:
    
    # Solicita ao usuário que insira um número através 
            # da função input.
    # A mensagem clara guia o usuário para inserir um número 
            # dentro do intervalo correto (1 a 7).
    # O input é imediatamente convertido para um inteiro 
            # usando int(), preparando-o para processamento.
    numero_digitado = int(input("Digite um número de 1 a 7: "))
    
    # Invoca a função 'dia_da_semana' passando o número 
            # digitado como argumento.
    # A função retorna o dia da semana correspondente ou 
            # uma mensagem de erro se o número for inválido.
    resultado = dia_da_semana(numero_digitado)
    
    # Exibe o resultado para o usuário, proporcionando 
            # feedback imediato sobre a entrada fornecida.
    print(resultado)

# Captura exceções de ValueError que podem ocorrer 
        # durante a conversão de input para int.
# Esse tratamento é essencial para lidar com entradas que 
        # não são numéricas, como letras ou símbolos,
        # garantindo que o programa não termine inesperadamente e 
        # forneça uma mensagem de erro compreensível.
except ValueError:
    
    # Informa ao usuário que a entrada foi inválida e 
            # solicita que apenas números sejam inseridos,
            # proporcionando uma experiência de usuário clara e instrutiva.
    print("Entrada inválida! Por favor, insira um número.")
```

```python
"""
3 - Faça um Programa que peça uma nota, entre zero e dez.

O programa deverá seguir as seguintes regras:

- Solicitar ao usuário que insira uma nota entre 0 e 10.
- Verificar se o valor inserido é válido, ou seja, se está dentro 
            do intervalo de 0 a 10, incluindo os extremos.
- Se o valor inserido for inválido (menor que 0, maior que 10 ou não for numérico), o 
            programa deverá exibir uma mensagem de erro e continuar pedindo a nota 
            até que o usuário insira um valor válido.
- Quando um valor válido for inserido, o programa deverá encerrar e exibir 
            uma mensagem confirmando a nota válida inserida.
"""

# Definindo uma função chamada `solicitar_nota` para encapsular o 
        # processo de solicitação e validação da nota.
def solicitar_nota():
    
    # Utilização de um laço `while True` para repetir indefinidamente 
            # até que um break seja chamado.
    while True:
        
        try:
            # A função `input` é usada para pedir ao usuário que digite 
                    # uma nota. O texto entre aspas é a mensagem exibida ao usuário.
            # `float()` é usado para converter a entrada do usuário em 
                    # um número de ponto flutuante, permitindo valores decimais.
            nota = float(input("Digite uma nota entre 0 e 10: "))
            
            # Condicional `if` que verifica se a `nota` está 
                    # entre 0 e 10, incluindo ambos os limites.
            if 0 <= nota <= 10:
                
                # Se a condição for verdadeira, a função retorna a nota, 
                        # saindo do laço e encerrando a função.
                return nota
                
            else:
                
                # Caso a nota não esteja dentro do intervalo 
                        # permitido, esta mensagem é impressa.
                print("Nota inválida! Por favor, insira um valor entre 0 e 10.")
        
        # O bloco `except` captura um erro específico, `ValueError`, 
                # que ocorre se a conversão para `float` falhar.
        except ValueError:
            
            # Se um erro `ValueError` ocorrer, esta mensagem é exibida.
            print("Entrada inválida! Por favor, insira um número.")

# O 'programa principal' começa aqui.
# A função `solicitar_nota` é chamada e seu retorno (uma nota válida) é 
        # armazenado na variável `nota_valida`.
nota_valida = solicitar_nota()

# A nota válida é exibida utilizando uma f-string, que permite 
        # inserir valores de variáveis diretamente no texto da string.
print(f"Você inseriu uma nota válida: {nota_valida}")
```

```python
"""
4 - Faça um Programa que leia uma quantidade indeterminada de números 
        positivos e conte quantos deles estão nos seguintes intervalos:

[0-25]
[26-50]
[51-75]
[76-100]

- O programa deverá seguir as seguintes regras:

- Solicitar que o usuário insira números positivos.

- Classificar cada número inserido dentro dos intervalos:

[0-25]
[26-50]
[51-75]
[76-100]

A entrada de números deve continuar até que o usuário insira um número negativo. 

Ao inserir um número negativo, a leitura deve ser encerrada, e o programa 
        deve exibir a contagem de números em cada intervalo.
        
Se o usuário inserir um número maior que 100, ele será ignorado e não 
        contará para nenhum dos intervalos.
            
Ao final, exibir quantos números pertencem a cada um dos intervalos.
"""

# Função para contar os números dentro dos intervalos especificados
def contar_numeros():
    
    # Inicializa contadores para cada intervalo
    # Essas variáveis servirão para armazenar a quantidade de 
            # números que se enquadram em cada um dos intervalos.
    intervalo_0_25 = 0
    intervalo_26_50 = 0
    intervalo_51_75 = 0
    intervalo_76_100 = 0

    # Loop infinito que continua solicitando números até 
            # que seja inserido um número negativo.
    while True:
        
        try:
            
            # Solicita ao usuário um número inteiro. O texto entre 
                    # aspas é a mensagem exibida ao usuário.
            # `int()` é utilizado para converter a entrada do 
                    # usuário (string) em um número inteiro.
            numero = int(input("Digite um número (negativo para encerrar): "))

            # Verifica se o número inserido é negativo.
            # Se for negativo, o loop `while` é interrompido usando `break`, e 
                    # o programa prossegue para a exibição dos resultados.
            if numero < 0:
                break

            # Verifica em qual dos intervalos definidos o número se encaixa.
            # Cada condição `if` ou `elif` compara o número para ver se 
                    # ele está dentro de um intervalo específico.
            if 0 <= numero <= 25:
                intervalo_0_25 += 1  # Incrementa o contador para o intervalo [0-25]
            elif 26 <= numero <= 50:
                intervalo_26_50 += 1  # Incrementa o contador para o intervalo [26-50]
            elif 51 <= numero <= 75:
                intervalo_51_75 += 1  # Incrementa o contador para o intervalo [51-75]
            elif 76 <= numero <= 100:
                intervalo_76_100 += 1  # Incrementa o contador para o intervalo [76-100]
            else:
                
                # Se o número for maior que 100, ele será ignorado.
                # Essa mensagem informa ao usuário que o número foi ignorado.
                print(f"O número {numero} está fora do intervalo considerado (0-100). Ignorado.")

        # Captura o erro se o usuário não inserir um valor inteiro.
        # Se `int()` falhar na conversão, é gerada uma exceção `ValueError`.
        except ValueError:
            
            # A mensagem é exibida para o usuário, pedindo que 
                    # insira um número inteiro válido.
            print("Entrada inválida! Por favor, insira um número inteiro.")

    # Após o encerramento do loop, as contagens de números em 
            # cada intervalo são exibidas.
    # Utiliza-se múltiplas linhas de `print()` para mostrar 
            # claramente as contagens para o usuário.
    print("\nContagem final de números em cada intervalo:")
    print(f"Intervalo [0-25]: {intervalo_0_25} número(s)")
    print(f"Intervalo [26-50]: {intervalo_26_50} número(s)")
    print(f"Intervalo [51-75]: {intervalo_51_75} número(s)")
    print(f"Intervalo [76-100]: {intervalo_76_100} número(s)")

# Programa principal
# Chama a função `contar_numeros()` para iniciar o processo.
contar_numeros()
```

```python
"""
5 - Faça um Programa que leia um vetor de 10 números reais e os mostre na ordem inversa.


O programa deve seguir as seguintes regras:

- Solicitar que o usuário insira 10 números reais.
- Armazenar os números inseridos em um vetor (ou lista).
- Após o usuário inserir todos os 10 números, o programa deve 
            exibir os números na ordem inversa em que foram inseridos.
- Certificar-se de que os números podem ser reais (incluindo 
        números com casas decimais).
"""

# Definição de uma função chamada `ler_vetor_inverso`.
# Esta função é responsável por solicitar 10 números reais ao 
        # usuário e exibir esses números na ordem inversa.
def ler_vetor_inverso():
    
    # Cria uma lista vazia chamada `vetor`.
    # Esta lista será usada para armazenar os 10 números 
            # reais digitados pelo usuário.
    vetor = []

    # Loop `for` que executa 10 vezes, uma para cada número 
            # que precisa ser inserido.
    # `range(10)` gera uma sequência de números de 0 a 9, e `i` é 
            # a variável que toma cada um desses valores na sequência.
    for i in range(10):
        
        # Loop interno `while True` que garante que cada número 
                # seja inserido corretamente.
        # Este loop só terminará quando um número real válido for 
                # digitado e adicionado à lista.
        while True:
            
            try:
                
                # Solicita ao usuário que digite um número real. Utiliza-se `input()` 
                        # para receber a entrada como string.
                # `float()` tenta converter essa string em um número 
                        # real (com casas decimais).
                # `f"Digite o {i+1}º número real: "` é uma f-string que 
                        # incorpora `i+1` para mostrar o número da entrada atual de 1 a 10.
                numero = float(input(f"Digite o {i+1}º número real: "))

                # Adiciona o número real digitado à lista `vetor`.
                # `append(numero)` adiciona o número ao final da lista.
                vetor.append(numero)

                # `break` é usado para sair do loop `while` após 
                        # uma entrada bem-sucedida.
                break
                
            except ValueError:
                
                # `except ValueError` captura erros que ocorrem se `float()` 
                        # não consegue converter a entrada em um número real.
                # Por exemplo, se o usuário digitar texto em vez de números.
                # A mensagem "Entrada inválida! Por favor, insira um 
                        # número real." é exibida se um erro for capturado.
                print("Entrada inválida! Por favor, insira um número real.")

    # Após coletar todos os 10 números, a lista é exibida 
            # na ordem inversa.
    # `print("\nOs números na ordem inversa são:")` simplesmente 
            # introduz a saída que será listada a seguir.
    print("\nOs números na ordem inversa são:")
    
    # `reversed(vetor)` retorna um iterador que acessa a 
            # lista `vetor` do último elemento ao primeiro.
    # O loop `for` itera sobre esse iterador, exibindo cada número.
    for numero in reversed(vetor):
        
        # `print(numero)` exibe cada número na ordem 
                # inversa em que foram inseridos.
        print(numero)

# Chamada à função `ler_vetor_inverso()` que inicia o processo.
# Esta linha é o ponto de entrada do programa, onde a 
        # função é chamada e a execução começa.
ler_vetor_inverso()
```

```python
"""
6 - Faça um Programa que leia 20 números inteiros e armazene-os num vetor.

O programa deve seguir as seguintes regras:

- Solicitar que o usuário insira 20 números inteiros.
- Armazenar todos os números em um vetor principal.
- Separar os números pares e armazená-los em um vetor chamado PAR.
- Separar os números ímpares e armazená-los em um vetor chamado IMPAR.

Após o usuário inserir os 20 números, o programa deve exibir os três vetores:
        - O vetor principal contendo todos os números inseridos.
        - O vetor PAR contendo apenas os números pares.
        - O vetor IMPAR contendo apenas os números ímpares.
"""

# Definição da função `separar_pares_impares`.
# Esta função é responsável por solicitar 20 números inteiros ao 
        # usuário e classificá-los em pares e ímpares.
def separar_pares_impares():
    
    # Inicialização de três listas:
    # `todos_numeros` armazenará todos os 20 números inseridos pelo usuário.
    # `pares` armazenará apenas os números pares.
    # `impares` armazenará apenas os números ímpares.
    todos_numeros = []
    pares = []
    impares = []

    # Loop `for` que itera 20 vezes, correspondendo aos 20 
            # números que o usuário deve inserir.
    for i in range(20):
        
        # Loop `while True` garante que a entrada seja um número 
                # inteiro válido antes de prosseguir.
        while True:
            
            try:
                
                # Solicitação de um número inteiro ao usuário. Utiliza-se `input()` 
                        # para obter a entrada como string.
                # `int()` converte a string para um inteiro. A f-string `f"Digite o {i+1}º 
                        # número inteiro: "` é usada para indicar qual 
                        # número está sendo pedido (de 1 a 20).
                numero = int(input(f"Digite o {i+1}º número inteiro: "))

                # Adiciona o número inteiro à lista `todos_numeros`.
                todos_numeros.append(numero)

                # Condicional para determinar se o número é par ou ímpar:
                # `% 2` calcula o resto da divisão do número por 2.
                # Se o resto é 0, o número é par, então é adicionado à lista `pares`.
                # Se o resto é diferente de 0, o número é ímpar, 
                        # então é adicionado à lista `impares`.
                if numero % 2 == 0:
                    pares.append(numero)
                else:
                    impares.append(numero)

                # `break` finaliza o loop `while`, pois um número 
                        # válido foi inserido.
                break
                
            except ValueError:
                
                # Mensagem de erro é exibida se `int()` falhar na 
                        # conversão da entrada em um número inteiro.
                print("Entrada inválida! Por favor, insira um número inteiro.")

    # Após a inserção e classificação dos números, as listas são exibidas:
    # Imprime a lista `todos_numeros` que contém todos os números inseridos.
    print("\nTodos os números:", todos_numeros)
    
    # Imprime a lista `pares` que contém apenas os números pares.
    print("Números pares:", pares)
    
    # Imprime a lista `impares` que contém apenas os números ímpares.
    print("Números ímpares:", impares)

# Chamada à função `separar_pares_impares` que 
        # inicia a execução do programa.
# Essa chamada é o ponto de entrada da 
        # funcionalidade da função.
separar_pares_impares()
```

```python
"""
7 - Faça um Programa que peça as quatro notas de 10 alunos, calcule e 
        armazene num vetor a média de cada aluno e imprima o número de
        alunos com média maior ou igual a 7.0.

O programa deve seguir as seguintes regras:

- Solicitar que o usuário insira quatro notas para cada um dos 10 alunos.
- Calcular a média de cada aluno e armazená-la num vetor.
- Contar quantos alunos têm média maior ou igual a 7.0.
- Ao final, exibir a quantidade de alunos com média maior ou igual a 7.0.
"""

# Função chamada `calcular_medias` responsável por calcular e 
        # armazenar as médias dos alunos e contar quantos têm média alta.
def calcular_medias():
    
    # Lista `medias` para armazenar a média de cada um dos 10 alunos.
    medias = []
    
    # Contador para os alunos que têm média maior ou igual a 7.0.
    alunos_com_media_maior_que_sete = 0

    # Loop externo que se repete 10 vezes, uma para cada aluno.
    for i in range(10):
        
        # Mensagem para o usuário indicando de qual aluno 
                # as notas serão inseridas.
        print(f"\nInsira as 4 notas do aluno {i+1}:")
        soma_notas = 0  # Variável para somar as notas de cada aluno.

        # Loop interno que se repete 4 vezes, uma para cada nota do aluno.
        for j in range(4):
            
            # Loop 'while True' garante a captura e validação 
                    # correta de cada nota inserida.
            while True:
                
                try:
                    
                    # Solicita ao usuário que insira uma nota.
                    nota = float(input(f"Digite a {j+1}ª nota: "))
                    
                    # Valida se a nota está no intervalo de 0 a 10.
                    if 0 <= nota <= 10:
                    
                        soma_notas += nota  # Soma a nota válida ao 
                                # total de notas do aluno.
                        break  # Sai do loop 'while' após uma entrada válida.
                    
                    else:
                    
                        # Mensagem de erro se a nota não estiver no intervalo válido.
                        print("Nota inválida! A nota deve estar entre 0 e 10.")
                        
                except ValueError:
                    
                    # Captura erros se a entrada não puder ser 
                            # convertida para float.
                    print("Entrada inválida! Por favor, insira um número real.")

        # Cálculo da média dividindo a soma total das notas por 4.
        media = soma_notas / 4
        
        # Adiciona a média calculada à lista `medias`.
        medias.append(media)

        # Verifica se a média é maior ou igual a 7.0.
        if media >= 7.0:
            
            # Incrementa o contador se a média for maior ou igual a 7.0.
            alunos_com_media_maior_que_sete += 1

    # Após coletar e calcular todas as médias, exibe quantos 
            # alunos têm média maior ou igual a 7.0.
    print(f"\nNúmero de alunos com média maior ou igual a 7.0: {alunos_com_media_maior_que_sete}")

# Chama a função `calcular_medias` para executar o programa.
# Esta é a parte do código que inicia a execução 
        # do processo descrito acima.
calcular_medias()
```

```python
"""
8 - Construa uma função que receba uma data no formato DD/MM/AAAA e 
        devolva uma string no formato D de mêsPorExtenso de AAAA.

Se a data for inválida (por exemplo, se o mês ou o dia não existirem), a 
        função deverá retornar NULL.
        
A função deve considerar os meses por extenso como: "Janeiro", "Fevereiro", etc.
"""

from datetime import datetime  # Importa a classe datetime do 
                               # módulo datetime para manipulação de datas.

# Dicionário para mapear os números dos meses aos 
        # seus respectivos nomes em português.
meses_por_extenso = {
    1: "Janeiro", 2: "Fevereiro", 3: "Março", 4: "Abril",
    5: "Maio", 6: "Junho", 7: "Julho", 8: "Agosto",
    9: "Setembro", 10: "Outubro", 11: "Novembro", 12: "Dezembro"
}

# Definição da função 'data_por_extenso' que aceita uma 
        # string 'data_str' representando uma data.
def data_por_extenso(data_str):
    
    try:
        
        # A função `strptime` tenta converter a string 'data_str' 
                # em um objeto datetime.
        # O formato "%d/%m/%Y" especifica que a string 
                # deve estar no formato dia/mês/ano.
        data = datetime.strptime(data_str, "%d/%m/%Y")
        
        # Atribui o dia, o mês e o ano às variáveis 'dia', 'mes' e 
                # 'ano', respectivamente.
        dia = data.day
        mes = data.month
        ano = data.year

        # Constrói a string no formato "D de mêsPorExtenso 
                # de AAAA" usando o dicionário 'meses_por_extenso'
                # para converter o número do mês em seu nome por extenso.
        return f"{dia} de {meses_por_extenso[mes]} de {ano}"

    except ValueError:
        
        # O bloco 'except' captura o ValueError, que ocorre 
                # se 'data_str' não corresponder ao formato esperado
                # ou representar uma data inexistente (como 31 de fevereiro).
        # Retorna "NULL" para indicar que a data é inválida.
        return "NULL"

# Programa principal que testa a função com dois exemplos.
data_teste_1 = "10/10/2024"  # Data válida
data_teste_2 = "31/02/2020"  # Data inválida (Fevereiro geralmente não tem 31 dias)

# Exibe os resultados das funções chamadas 
        # com as datas de teste.
print(data_por_extenso(data_teste_1))  # Deve retornar: 10 de Outubro de 2024
print(data_por_extenso(data_teste_2))  # Deve retornar: NULL
```

```python
"""
10. Leonardo o Pescador comprou um microcomputador para controlar o rendimento 
        diário de seu trabalho. 
Sempre que ele traz um peso de peixes maior que o estabelecido pelo 
        regulamento de pesca do estado de São Paulo (50 quilos), 
        ele deve pagar uma multa de R$ 8,00 por quilo excedente.

O programa deve seguir as seguintes regras:

- Solicitar que o usuário insira o peso total de peixes pescados em quilos.
- Se o peso for maior que 50 quilos, calcular o excesso e o valor da multa.
- Armazenar a quantidade de quilos excedentes na variável excesso.
- Armazenar o valor da multa na variável multa (R$ 8,00 por quilo excedente).
- Exibir o peso total, o excesso (se houver) e o valor da multa (se houver).
"""

# Função para calcular o excesso de peso e a multa
def calcular_multa(peso):
    
    # Define o limite de peso permitido sem multa, de 
            # acordo com a regulamentação.
    limite = 50
    
    # Define o valor da multa por quilo excedente, conforme 
            # especificado no problema.
    valor_multa_por_quilo = 8.00

    # Verifica se o peso dos peixes é maior que o limite permitido.
    if peso > limite:
        
        # Calcula quanto o peso excede o limite.
        excesso = peso - limite
        
        # Calcula o valor total da multa com base no peso excedente.
        multa = excesso * valor_multa_por_quilo
        
    else:
        
        # Se não houver excesso, define excesso e multa como zero.
        excesso = 0
        multa = 0

    # Retorna os valores de excesso e multa para serem 
            # usados fora da função.
    return excesso, multa

# Programa principal
try:
    
    # Solicita ao usuário que digite o peso total dos 
            # peixes pescados em quilos.
    peso = float(input("Digite o peso total de peixes (em quilos): "))
    
    # Chama a função `calcular_multa` passando o peso 
            # inserido e obtém o excesso e o valor da multa.
    excesso, multa = calcular_multa(peso)
    
    # Exibe o peso total dos peixes pescados.
    print(f"\nPeso total de peixes: {peso:.2f} kg")
    
    # Exibe o excesso de peso, se houver.
    print(f"Excesso de peso: {excesso:.2f} kg")
    
    # Exibe o valor da multa, se houver.
    print(f"Valor da multa: R$ {multa:.2f}")

except ValueError:
    
    # Captura erro se a entrada não for um número e informa ao 
            # usuário para inserir um valor numérico válido.
    print("Entrada inválida! Por favor, insira um valor numérico válido.")
```

```python
"""
11. Faça um programa para uma loja de tintas.

O programa deve seguir as seguintes regras:

- Solicitar ao usuário o tamanho da área a ser pintada em metros quadrados.
- Considerar que a cobertura da tinta é de 1 litro para cada 5 metros quadrados.
- A tinta é vendida em latas de 18 litros, e cada lata custa R$ 100,00.

O programa deve calcular:
- A quantidade de latas de tinta necessárias para cobrir a área.
- O preço total das latas de tinta.
- Informe ao usuário a quantidade de latas a serem compradas e o preço total.
"""

# Importa a biblioteca 'math' para usar a função 'ceil', 
        # que arredonda para cima.
import math  

# Função para calcular a quantidade de latas de 
        # tinta e o preço total
def calcular_tinta(area):
    
    # Define a cobertura que cada litro de tinta oferece, 
            # que é de 5 metros quadrados por litro.
    cobertura_por_litro = 5

    # Define a quantidade de litros por lata de tinta, 
            # que é de 18 litros por lata.
    litros_por_lata = 18
    
    # Define o preço de cada lata de tinta.
    preco_por_lata = 100.00

    # Calcula quantos litros de tinta são necessários 
            # para pintar a área informada.
    # Isso é feito dividindo a área pela cobertura que 
            # cada litro de tinta proporciona.
    litros_necessarios = area / cobertura_por_litro

    # Calcula a quantidade de latas necessárias, arredondando para cima.
    # 'math.ceil' é usado porque mesmo que a cobertura exceda 
            # apenas um pouco a capacidade de uma lata,
            # ainda será necessária uma lata inteira adicional.
    latas_necessarias = math.ceil(litros_necessarios / litros_por_lata)

    # Calcula o preço total, multiplicando o número de 
            # latas necessárias pelo preço de cada lata.
    preco_total = latas_necessarias * preco_por_lata

    # Retorna a quantidade de latas necessárias e o 
            # preço total como uma tupla.
    return latas_necessarias, preco_total

# Programa principal que solicita entrada do 
        # usuário e exibe os resultados.
try:
    
    # Pede ao usuário para digitar o tamanho da área a 
            # ser pintada em metros quadrados.
    area = float(input("Digite o tamanho da área a ser pintada (em metros quadrados): "))
    
    # Verifica se a área inserida é válida (maior que zero).
    if area <= 0:
        
        print("A área deve ser maior que zero!")
        
    else:
        
        # Se a área for válida, chama a função 'calcular_tinta' 
                # passando a área como argumento.
        latas, preco = calcular_tinta(area)

        # Exibe a quantidade de latas necessárias e o preço total 
                # usando formatação de strings para manter 
                # duas casas decimais no preço.
        print(f"\nQuantidade de latas de tinta necessárias: {latas} lata(s)")
        print(f"Preço total: R$ {preco:.2f}")
        

except ValueError:
    
    # Trata erro caso a entrada não seja um número válido.
    print("Entrada inválida! Por favor, insira um valor numérico válido.")
```

```python
"""
13. Crie um programa que:

- Solicite ao usuário que insira o tamanho desejado para a senha.
- Gere uma senha aleatória de acordo com o tamanho informado.
- A senha pode conter letras maiúsculas, minúsculas, números e 
            caracteres especiais.
"""

# Importando módulos necessários:
import random  # Módulo para gerar dados aleatórios.
import string  # Módulo que contém sequências de caracteres 
        # comuns como letras e dígitos.

# Definição da função 'gerar_senha' que aceita um parâmetro: 'tamanho'.
def gerar_senha(tamanho):
    
    # Concatena strings de caracteres alfabéticos (maiúsculas e 
            # minúsculas), numéricos e símbolos.
    # 'string.ascii_letters' inclui todas as letras 
            # de A a Z (maiúsculas e minúsculas).
    # 'string.digits' inclui todos os dígitos de 0 a 9.
    # 'string.punctuation' inclui todos os caracteres de 
            # pontuação como !, @, #, etc.
    caracteres = string.ascii_letters + string.digits + string.punctuation
    
    # Gera a senha usando uma compreensão de lista:
    # 'random.choice(caracteres)' seleciona um caractere 
            # aleatório da string 'caracteres'.
    # 'for i in range(tamanho)' repete a seleção 'tamanho' 
            # vezes para formar uma senha do comprimento desejado.
    senha = ''.join(random.choice(caracteres) for i in range(tamanho))
    
    # Retorna a senha gerada.
    return senha

# Bloco principal do programa, executado ao iniciar o script.
try:
    
    # Solicita ao usuário para digitar o tamanho da 
            # senha que deseja.
    # 'int(input(...))' converte a entrada do usuário para um 
            # inteiro, necessário para o loop.
    tamanho_senha = int(input("Digite o tamanho desejado para a senha: "))
    
    # Verifica se o tamanho da senha é ao menos 1, já que não é 
            # possível ter uma senha de tamanho zero ou negativo.
    if tamanho_senha < 1:

        # Aviso caso o tamanho seja inapropriado.
        print("O tamanho da senha deve ser maior que zero!")  
        
    else:
        
        # Chama a função 'gerar_senha' com o tamanho fornecido e 
                # armazena o resultado em 'senha_gerada'.
        senha_gerada = gerar_senha(tamanho_senha)
        
        # Exibe a senha gerada.
        print(f"Senha gerada: {senha_gerada}")

except ValueError:
    
    # Captura erro caso a conversão para inteiro falhe, o que 
            # acontece se o usuário não digitar um número.
    print("Entrada inválida! Por favor, insira um número inteiro.")
```

```python
"""
14. Escreva um programa que:

- Receba do usuário uma série de nomes em uma única linha, separados por espaços.
- Organize esses nomes em ordem alfabética.
- Exiba os nomes ordenados em uma única linha, separados por espaço.
"""

# Este é um programa compacto que faz tudo em uma linha. 
        # Vamos quebrar cada parte:

# 1. `input()`: Esta função é chamada para coletar uma entrada do usuário. 
        # Espera-se que o usuário digite uma série de nomes separados por espaços.
# Por exemplo, o usuário pode digitar: "Ana Carlos Bruno Diana".

# 2. `.split()`: Este método é aplicado à string que `input()` retorna. 
        # Ele divide a string em uma lista de substrings.
# O método `.split()` sem argumentos divide a string por 
        # espaços em branco. 
# Assim, a string "Ana Carlos Bruno Diana" se torna a 
        # lista ['Ana', 'Carlos', 'Bruno', 'Diana'].

# 3. `sorted(...)`: Esta função recebe a lista de nomes e retorna 
        # uma nova lista com os elementos ordenados alfabeticamente.
# Portanto, se a entrada for ['Ana', 'Carlos', 'Bruno', 'Diana'], o 
        # resultado de `sorted()` será ['Ana', 'Bruno', 'Carlos', 'Diana'].

# 4. `print(*...)`: O asterisco `*` antes de uma lista no `print()` é 
        # usado para "desempacotar" os elementos da lista. 
# Cada elemento é passado como um argumento separado 
        # para a função `print()`.
# Isso significa que em vez de imprimir a lista como um 
        # objeto único, `print()` exibirá cada elemento da 
        # lista separado por um espaço. 
# Assim, em vez de imprimir "['Ana', 'Bruno', 'Carlos', 'Diana']", 
        # ele imprime "Ana Bruno Carlos Diana".

# Aqui está o código completo em uma linha, que é elegante e 
        # eficiente para a tarefa de organizar e exibir nomes:

print(*sorted(input().split()))
```

```python
"""
17. Desenvolva um programa em Python que receba do usuário uma
        lista de números diferentes e retorne o maior deles. 

O programa deve incluir as seguintes funcionalidades adicionais:

- Funções Aninhadas: O programa deve utilizar funções aninhadas para 
        realizar as comparações entre os valores e determinar 
        qual é o maior número.

- Trabalho com Quantidade Variável de Números: O programa deve permitir 
        que o usuário insira qualquer quantidade de números (não 
        limitado a três) separados por espaço.

- Verificação de Duplicidade: Caso o usuário insira números duplicados, o 
        programa deve notificar o usuário e solicitar a entrada 
        de números diferentes.

- Ordenação Decrescente: Após encontrar o maior valor, o programa 
        deve exibir todos os números fornecidos pelo usuário em 
        ordem decrescente.

- Reinicialização Automática: O programa deve permitir ao usuário escolher 
        se deseja rodá-lo novamente ao final da execução, sem 
        precisar reiniciar manualmente.

"""

def comparar_dois_numeros(a, b):
    """
    Compara dois números e retorna o maior.
    
    Esta função é projetada para aceitar dois argumentos numéricos, 'a' e 'b', 
        e retorna o maior entre eles. Ela simplifica a tarefa de determinar o maior 
        número em várias partes de um programa.
    
    Parâmetros:
    - a (float|int): O primeiro número para comparação.
    - b (float|int): O segundo número para comparação.
    
    Retorna:
    - float|int: O maior dos dois números.
    """

    # Avalia se o número 'a' é maior que o número 'b'.
    if a > b:
        
        # Se 'a' for maior, retorna 'a' como o maior número.
        return a
        
    else:
        
        # Se 'a' não for maior, retorna 'b', pois 'b' é maior ou igual a 'a'.
        return b



def maior_valor(lista_numeros):
    
    """
    Determina o maior número em uma lista usando uma função 
            aninhada para comparações.

    A função inicia com o primeiro elemento da lista como o maior encontrado até então,
            e itera sobre o restante dos elementos para encontrar o maior
            número usando comparações de dois em dois elementos.

    Parâmetros:
            
            - lista_numeros (list of float|int): Lista dos 
                    números dos quais o maior será encontrado.
    
    Retorna:
            - float|int: O maior número encontrado na lista.
    """

    # Inicializa a variável 'maior' com o primeiro elemento da lista.
    # Esse elemento serve como ponto de partida para as comparações subsequentes.
    maior = lista_numeros[0]
    
    # Itera sobre os elementos da lista a partir do segundo elemento (índice 1).
    for num in lista_numeros[1:]:
        
        # Utiliza a função 'comparar_dois_numeros' para comparar o 
                # valor atual de 'maior'
                # com o próximo número na lista ('num').
        maior = comparar_dois_numeros(maior, num)
        # Atualiza 'maior' com o retorno da função 'comparar_dois_numeros', que é
                # o maior entre o 'maior' atual e 'num'.

    # Após a iteração completa, retorna o maior número encontrado na lista.
    return maior



def verificar_valores_duplicados(lista_numeros):
    
    """
    Verifica se existem valores duplicados na lista de números fornecida.

    Parâmetros:
    - lista_numeros (list of float|int): Lista de números a ser verificada.

    Retorna:
    - bool: Retorna True se houver duplicatas, False caso contrário.

    Esta função faz uso da propriedade de conjuntos (set) em Python, que não permite duplicatas,
        para determinar se a lista original contém valores duplicados.
    """

    # Compara o comprimento da lista original com o comprimento 
            # do conjunto criado a partir da lista.
    # O conjunto (set) removerá automaticamente quaisquer 
            # valores duplicados.
    # Se os comprimentos forem diferentes, isso indica a 
            # presença de duplicatas na lista original.
    if len(lista_numeros) != len(set(lista_numeros)):
        return True  # Retorna True indicando a presença de duplicatas.
    return False     # Retorna False se não houver duplicatas.


def solicitar_numeros():
    
    """
    Solicita ao usuário que insira uma lista de números, 
            garantindo que todos sejam diferentes.

    Retorna:
    
            - list of float: Retorna uma lista de números de 
                    ponto flutuante sem duplicatas.

    Esta função continua solicitando números até que uma entrada 
            válida (apenas números e sem duplicatas) seja fornecida.
    """

    # Inicia um loop infinito que só será quebrado manualmente 
            # se as condições corretas forem atendidas.
    while True:  
        
        try:
            
            # Solicita ao usuário que insira vários números, 
                    # esperando que sejam separados por espaço.
            entrada = input("Digite números diferentes separados por espaço: ")
            
            # Converte a entrada (uma string) em uma lista de 
                    # números de ponto flutuante.
            # 'split()' divide a string em uma lista onde cada 
                    # elemento é um número em forma de string.
            # 'map(float, ...)' converte cada elemento da lista 
                    # em um número de ponto flutuante.
            lista_numeros = list(map(float, entrada.split()))

            # Chama a função 'verificar_valores_duplicados' para 
                    # checar se há duplicatas na lista de números.
            if verificar_valores_duplicados(lista_numeros):
                
                # Se duplicatas forem encontradas, informa o 
                        # usuário e reinicia o loop.
                print("Os números devem ser diferentes. Tente novamente.")

                # Continua para a próxima iteração do loop.
                continue  

            # Retorna a lista de números se não houver duplicatas.
            return lista_numeros  
            
        except ValueError:
            
            # Captura um erro se a conversão para float falhar, o que 
                    # indica que a entrada não era composta somente por números.
            print("Entrada inválida. Digite apenas números.")


def exibir_numeros_ordenados(lista_numeros):
    
    """
    Exibe os números fornecidos pelo usuário em ordem decrescente.

    Parâmetros:
    - lista_numeros (list of float|int): Lista dos números que 
            serão ordenados e exibidos.

    Esta função ordena a lista de números em ordem decrescente e então os imprime,
            facilitando a visualização da disposição dos valores de maior para menor.
    """

    # Ordena a lista de números em ordem decrescente. 'sorted()' é 
            # uma função embutida que retorna uma nova lista ordenada.
    # O argumento 'reverse=True' especifica que a 
            # ordem deve ser decrescente.
    lista_ordenada = sorted(lista_numeros, reverse=True)

    # Imprime a lista ordenada formatada como uma string, 
            # mostrando claramente os números em ordem decrescente.
    print(f"Números em ordem decrescente: {lista_ordenada}")


def programa_principal():
    
    """
    Executa o programa principal, que solicita ao usuário uma 
            série de números, exibe o maior número,
            mostra todos os números em ordem decrescente e 
            permite a repetição do processo.

    Esta função coordena todas as ações, desde a entrada do usuário 
            até a exibição dos resultados,
            e gerencia a lógica de repetição do programa.
    """

    # Inicia um loop infinito, permitindo ao usuário repetir o 
            # processo de entrada e exibição de dados.
    while True:
        
        # Solicita uma lista de números ao usuário, garantindo 
                # que todos sejam diferentes e válidos.
        numeros = solicitar_numeros()

        # Utiliza a função 'maior_valor' para encontrar o maior 
                # número na lista fornecida.
        maior_numero = maior_valor(numeros)
        
        # Imprime o maior número encontrado, fornecendo feedback 
                # imediato ao usuário.
        print(f"O maior número é: {maior_numero}")

        # Chama a função 'exibir_numeros_ordenados' para mostrar 
                # todos os números em ordem decrescente.
        exibir_numeros_ordenados(numeros)

        # Pergunta ao usuário se deseja executar o programa novamente.
        continuar = input("Deseja rodar o programa novamente? (s/n): ").strip().lower()
        
        # Se o usuário digitar 'n', o loop é interrompido e 
                # o programa é encerrado.
        if continuar != 's':
            print("Programa encerrado.")
            break


# Executa o programa principal
programa_principal()
```

```python
"""
18. Exercício: Contagem de Letras Maiúsculas e Minúsculas

Escreva um programa em Python que leia uma palavra ou frase 
        digitada pelo usuário e retorne o número total de letras 
        maiúsculas e minúsculas presentes. O programa deve utilizar 
        apenas lógica básica e funções embutidas do Python para 
        realizar as verificações.

Requisitos:

- O programa deve solicitar que o usuário digite uma palavra ou frase.
- O programa deve contar o número de letras maiúsculas e o 
        número de letras minúsculas separadamente.
- Deve-se ignorar qualquer caractere que não seja uma letra, 
        como números, espaços, e pontuações.
- Utilize apenas funções embutidas do Python, como isalpha(), 
        isupper(), e islower() para realizar as verificações. 
"""

def contar_maiusculas_minusculas(frase):
    
    """
    Conta o número de letras maiúsculas e minúsculas em 
            uma frase dada pelo usuário.
    
    Argumentos:
            frase (str): A frase na qual as letras serão contadas.
    
    Retorna:
            int, int: Retorna dois inteiros representando o 
                    número de letras maiúsculas e minúsculas.
    """

    # Inicializa duas variáveis com zero para contar quantas 
            # letras maiúsculas e minúsculas estão na frase.
    maiusculas = 0
    minusculas = 0

    # O loop 'for' percorre cada caractere na string 
            # que o usuário inseriu.
    for caractere in frase:
        
        # Verifica se o caractere atual é alfabético, ignorando 
                # números, espaços e símbolos.
        if caractere.isalpha():  
            
            # Verifica se o caractere alfabético é uma letra maiúscula.
            if caractere.isupper():
                
                # Se for maiúscula, incrementa o contador de maiúsculas.
                maiusculas += 1  
                
            # Verifica se o caractere alfabético é uma letra minúscula.
            elif caractere.islower():
                
                # Se for minúscula, incrementa o contador de minúsculas.
                minusculas += 1  

    # Retorna os contadores de letras maiúsculas e 
            # minúsculas como uma tupla.
    return maiusculas, minusculas



def programa_principal():
    
    """
    Executa o programa principal que interage com o usuário e 
            mostra os resultados do processamento.
    
    Funções:
    - Solicita entrada de texto do usuário.
    - Utiliza a função `contar_maiusculas_minusculas` para processar o texto.
    - Exibe os resultados.
    """

    # Solicita ao usuário que digite uma palavra ou frase para análise.
    entrada = input("Digite uma palavra ou frase: ")

    # Chama a função `contar_maiusculas_minusculas` para obter a 
            # contagem de maiúsculas e minúsculas da entrada.
    total_maiusculas, total_minusculas = contar_maiusculas_minusculas(entrada)

    # Imprime o número de letras maiúsculas encontradas na entrada.
    print(f"Número de letras maiúsculas: {total_maiusculas}")
    
    # Imprime o número de letras minúsculas encontradas na entrada.
    print(f"Número de letras minúsculas: {total_minusculas}")

# Inicia a execução do programa principal quando o script é executado.
programa_principal()
```

```python
"""
21. Contagem de Segundas-Feiras no Dia 1º em um Intervalo de Datas

Escreva um programa em Python que solicita ao usuário um intervalo de 
        datas, com uma data de início e uma data de fim, e conta quantas 
        segundas-feiras caem no dia 1º de cada mês dentro desse intervalo. 
        
O programa deve validar as entradas e garantir que a data de início 
        seja anterior ou igual à data de fim.

Requisitos:
 - O programa deve solicitar que o usuário insira uma data de início e 
         uma data de fim no formato "AAAA-MM-DD".
- O programa deve verificar se o dia 1º de cada mês dentro do intervalo 
        dado é uma segunda-feira e contar quantas vezes isso ocorre.
- Se a data de início for posterior à data de fim, o programa deve 
        exibir uma mensagem de erro e solicitar que o usuário 
        insira novamente as datas.
- Caso o formato da data seja inválido, o programa deve solicitar a 
        correção até que o formato correto seja inserido.
"""

# Importando o módulo datetime, necessário para trabalhar com datas.
import datetime

# Definição da função que conta as segundas-feiras que caem 
        # no dia 1 de cada mês num intervalo de datas especificado.
def contar_segundas_feiras(data_inicio, data_fim):
    
    """
    Conta quantas segundas-feiras caem no dia 1º de cada mês 
            entre duas datas fornecidas.

    Parâmetros:
        - data_inicio (datetime.date): A data que inicia o 
                    intervalo de contagem.
        - data_fim (datetime.date): A data que termina o 
                    intervalo de contagem.

    Retorna:
        - int: O número de segundas-feiras que caem no dia 1º de 
                cada mês dentro do intervalo especificado.
    """

    # Inicialização do contador de segundas-feiras.
    contador_segundas = 0

    # Extração do ano e do mês das datas de início e fim.
    # Isso determina a partir de que ano e mês começar e onde terminar.
    ano_inicio, mes_inicio = data_inicio.year, data_inicio.month
    ano_fim, mes_fim = data_fim.year, data_fim.month

    # Loop que passa por cada ano no intervalo de anos desde o 
            # ano de início até o ano de fim.
    for ano in range(ano_inicio, ano_fim + 1):
        
        # Define o mês de início para o ano atual. Se for o 
                # primeiro ano do intervalo, começa do 'mes_inicio'.
        # Caso contrário, começa de janeiro (mes 1).
        mes_inicio_atual = mes_inicio if ano == ano_inicio else 1
        
        # Define o mês de fim para o ano atual. Se for o último 
                # ano do intervalo, termina no 'mes_fim'.
        # Caso contrário, termina em dezembro (mes 12).
        mes_fim_atual = mes_fim if ano == ano_fim else 12

        # Loop que passa por cada mês no ano atual, desde o 
                # mês de início até o mês de fim.
        for mes in range(mes_inicio_atual, mes_fim_atual + 1):
            
            # Criação de um objeto datetime.date para o primeiro 
                    # dia do mês corrente.
            # Isso é necessário para verificar o dia da semana 
                    # daquela data específica.
            data = datetime.date(ano, mes, 1)
            
            # Checagem se o primeiro dia do mês é uma segunda-feira.
            # 'weekday()' retorna 0 para segunda-feira, 1 para terça-feira, etc.
            if data.weekday() == 0:
                
                # Se for segunda-feira, incrementa o contador de segundas-feiras.
                contador_segundas += 1

    # Retorna o total de segundas-feiras encontradas que caem no 
            # primeiro dia dos meses no intervalo dado.
    return contador_segundas


# Definindo a função que solicita uma data do usuário.
def solicitar_data():
    
    """
    Solicita uma data do usuário e verifica se está no formato correto "AAAA-MM-DD".
    Se o formato estiver correto, retorna a data como um objeto datetime.date.
    Se estiver incorreto, solicita novamente.
    """
    
    # Inicia um loop infinito para continuar solicitando a data até 
            # que uma entrada válida seja recebida.
    while True:
        
        try:
            
            # Pede ao usuário para digitar uma data no formato especificado.
            data_str = input("Digite uma data (formato: AAAA-MM-DD): ")
            
            # Tenta converter a string de data em um objeto de data 
                    # utilizando strptime.
            # Se a string não estiver no formato "AAAA-MM-DD", uma 
                    # exceção ValueError será lançada.
            data = datetime.datetime.strptime(data_str, "%Y-%m-%d").date()
            
            # Se a data for convertida com sucesso, termina o loop 
                    # retornando o objeto de data.
            return data
            
        except ValueError:
            
            # Se um erro ValueError for capturado (formato errado), 
                    # informa o usuário e repete o loop.
            print("Data inválida. Tente novamente no formato AAAA-MM-DD.")

# Instrução para o usuário sobre o que será solicitado.
print("Por favor, insira o intervalo de datas.")

# Solicita ao usuário que forneça uma data de início 
        # utilizando a função 'solicitar_data'.
data_inicio = solicitar_data()

# Solicita ao usuário que forneça uma data de fim 
        # utilizando a mesma função.
data_fim = solicitar_data()


# Verifica se a data de início é anterior à data de fim para 
        # garantir que o intervalo seja válido.
if data_inicio > data_fim:
    
    # Se a data de início for posterior à data de fim, 
            # imprime uma mensagem de erro.
    print("A data de início deve ser anterior à data de fim.")
    
else:
    
    # Se a data de início for anterior ou igual à data de fim, 
            # procede com o cálculo.
    
    # Chama a função 'contar_segundas_feiras' passando as datas de 
            # início e fim como argumentos.
    # Esta função retorna o número de segundas-feiras que caem no 
            # primeiro dia de cada mês no intervalo especificado.
    resultado = contar_segundas_feiras(data_inicio, data_fim)
    
    # Imprime o resultado do cálculo, mostrando quantas segundas-feiras 
            # caem no dia 1º de cada mês no intervalo dado.
    print(f"O número de segundas-feiras que caem no dia 1º de cada mês entre {data_inicio} e {data_fim} é: {resultado}")
```

```python
"""
22. Reorganização de Carros por Ano de Fabricação

Escreva um programa em Python que receba uma lista de dicionários, 
        onde cada dicionário descreve um carro e contém as 
        seguintes informações:

- Modelo do carro (chave: "modelo").
- Ano de fabricação do carro (chave: "ano_fabricacao").
- O objetivo é reorganizar a lista de carros de acordo com o 
        ano de fabricação em ordem crescente, sem modificar a 
        lista original, e exibir a lista reorganizada.

Requisitos:
- A lista de carros deve ser composta por dicionários, onde 
        cada dicionário tem as chaves "modelo" e "ano_fabricacao".
- O programa deve reorganizar a lista com base no valor do 
        campo "ano_fabricacao", do mais antigo ao mais recente.
- A lista original deve permanecer inalterada, e o resultado deve 
        ser uma nova lista com os carros ordenados.
"""


# Função para reorganizar os carros por ano de fabricação, 
        # retornando uma nova lista ordenada.
def organizar_por_ano(carros):
    
    """
    Esta função recebe uma lista de dicionários que representam 
            carros e os ordena pelo ano de fabricação.
    A lista original não é modificada porque a função 'sorted' 
            cria uma nova lista ordenada.

    Parâmetros:
    carros (list): Lista de dicionários onde cada dicionário 
            contém informações sobre um carro.

    Retorna:
    list: Uma nova lista contendo os dicionários dos 
            carros ordenados pelo 'ano_fabricacao'.
    """
    
    # A função 'sorted' é usada para ordenar a lista. Ela é 
            # uma função embutida que ordena qualquer iterável.
    # Aqui usamos uma função lambda como chave de ordenação. 
    # 'lambda carro: carro['ano_fabricacao']' pega o valor do ano 
            # de fabricação de cada carro para a ordenação.
    return sorted(carros, key=lambda carro: carro['ano_fabricacao'])

# Lista inicial de carros. Cada carro é representado por um 
        # dicionário com chaves "modelo" e "ano_fabricacao".
carros = [
    {"modelo": "Fusca", "ano_fabricacao": 1980},
    {"modelo": "Civic", "ano_fabricacao": 2010},
    {"modelo": "Corolla", "ano_fabricacao": 2005},
    {"modelo": "Fiesta", "ano_fabricacao": 2008},
    {"modelo": "Gol", "ano_fabricacao": 2000}
]

# Imprimindo a lista original de carros para mostrar que a 
        # lista inicial permanece inalterada após a ordenação.
# Esta linha inicializa a exibição de conteúdo no console.
print("Lista original de carros:")

# Este loop 'for' passa por cada item na lista 'carros'. 
        # Cada item é um dicionário representando um carro.
for carro in carros:
    
    # A função 'print()' é usada para exibir cada carro. 
    # Aqui, 'carro' é um dicionário, e quando passado 
            # para 'print()', é convertido em uma string e exibido.
    print(carro)

# Chamando a função 'organizar_por_ano' para ordenar os 
        # carros por ano de fabricação.
# 'organizar_por_ano(carros)' chama a função que 
        # organiza a lista de carros
        # e retorna uma nova lista ordenada sem alterar a original.
carros_organizados = organizar_por_ano(carros)

# Imprimindo a nova lista de carros, que agora 
        # está organizada por ano de fabricação.
# Este print introduz a seção onde os carros 
        # ordenados serão mostrados.
print("\nLista de carros organizada por ano de fabricação:")

# Este loop 'for' itera sobre a lista 'carros_organizados'.
# A cada iteração, 'carro' recebe um dicionário da lista ordenada.
for carro in carros_organizados:
    
    # 'print(carro)' exibe o dicionário de cada carro, 
            # que agora está ordenado pelo ano de fabricação.
    print(carro)
```

```python
"""
23. Diferentes Formas de Somar os Elementos de uma Lista

Dada a lista x = [5, 10, 15, 20], apresente pelo menos cinco formas diferentes 
        de retornar a soma dos elementos dessa lista. Utilize diferentes 
        abordagens e estruturas de controle ou funções nativas do Python.

Requisitos:

- Apresente cinco maneiras distintas de calcular e exibir a 
        soma dos elementos da lista.
- As abordagens devem variar entre o uso de funções embutidas, laços de 
        repetição, compreensões de lista e outras ferramentas da linguagem Python.
- O resultado final para cada método deve ser o mesmo: a soma dos elementos da lista. 
"""

# Método 1: Usando a função embutida sum()
# Definição da lista de números a serem somados.
x = [5, 10, 15, 20]

# Uso da função sum() para calcular a soma dos 
        # elementos da lista x.
soma = sum(x)

# Impressão do resultado da soma.
print("Soma usando sum():", soma)

# Método 2: Usando um laço for para somar os elementos
# Definição da lista de números a serem somados.
x = [5, 10, 15, 20]

# Inicialização de uma variável para acumular a soma dos elementos.
soma = 0

# Laço for para iterar sobre cada número na lista x e acumular a soma.
for num in x:
    soma += num
    
# Impressão do resultado da soma.
print("Soma usando for loop:", soma)

# Método 3: Usando a função reduce() da biblioteca 
        # functools para somar elementos
# Importação da função reduce do módulo functools.
from functools import reduce

# Definição da lista de números a serem somados.
x = [5, 10, 15, 20]

# Uso da função reduce com uma função lambda que soma dois 
        # elementos, aplicada a todos os elementos da lista x.
soma = reduce(lambda a, b: a + b, x)

# Impressão do resultado da soma.
print("Soma usando reduce():", soma)

# Método 4: Usando um laço while para somar os elementos
# Definição da lista de números a serem somados.
x = [5, 10, 15, 20]

# Inicialização de uma variável para acumular a soma e 
        # outra para servir como índice.
soma = 0
i = 0

# Laço while que continua enquanto o índice i for menor 
        # que o número de elementos na lista x.
while i < len(x):

    # Soma o elemento na posição i da lista ao acumulador.
    soma += x[i]  

    # Incremento do índice i.
    i += 1  
    
# Impressão do resultado da soma.
print("Soma usando while loop:", soma)

# Método 5: Usando list comprehension e a função sum() 
        # para somar elementos.
# Definição da lista de números a serem somados.
x = [5, 10, 15, 20]

# Uso de list comprehension para criar uma nova lista com os 
        # mesmos elementos de x e soma desses elementos 
        # com a função sum().
soma = sum([num for num in x])

# Impressão do resultado da soma.
print("Soma usando list comprehension:", soma)
```

```python
"""
24. Implemente um sistema de carrinho de compras simples em Python. 

O sistema deverá permitir que o usuário adicione e remova itens do carrinho, 
bem como visualize o conteúdo do carrinho. A interação com o sistema será feita 
via comandos de texto e ele deve permanecer ativo até que o usuário escolha encerrar.

O sistema deve seguir estas instruções:

Comandos:

- adicionar: Solicita o nome de um item e o adiciona ao carrinho.
- remover: Solicita o nome de um item e o remove do carrinho, caso ele esteja presente.
- ver: Exibe o conteúdo atual do carrinho.
- sair: Encerra o programa.

Regras:

- Se o usuário tentar remover um item que não está no carrinho, 
        uma mensagem apropriada deve ser exibida.
- Se o usuário tentar remover itens de um carrinho vazio, uma 
        mensagem informativa também deve ser exibida.
- O sistema deve seguir funcionando até que o comando sair seja digitado.

Restrições:

- Não utilize compreensão de listas ou funções que retornem listas vazias.
- Apenas estruturas condicionais simples (if, else, elif) devem ser usadas.
"""

# Inicialização do carrinho como uma lista vazia, 
        # onde os itens serão armazenados
carrinho = []

# Início do loop principal do sistema, que mantém o sistema 
        # ativo até que o usuário decida sair
while True:
    
    # Solicita ao usuário que digite um comando e converte o 
            # texto para minúsculas para padronização
    comando = input("Digite 'adicionar' para incluir um item, 'remover' para excluir, 'ver' para exibir o carrinho ou 'sair' para encerrar: ").lower()

    # Verifica se o comando digitado é 'adicionar'
    if comando == 'adicionar':
        
        # Solicita ao usuário o nome do item a ser adicionado ao carrinho
        item = input("Digite o nome do item que deseja adicionar: ")
        
        # Adiciona o item à lista carrinho
        carrinho.append(item)
        
        # Exibe uma mensagem confirmando que o item foi adicionado
        print(f"'{item}' adicionado ao carrinho.")

    # Verifica se o comando digitado é 'remover'
    elif comando == 'remover':
        
        # Verifica se há itens no carrinho antes de tentar remover
        if carrinho:
            
            # Solicita ao usuário o nome do item a ser removido
            item = input("Digite o nome do item que deseja remover: ")
            
            # Verifica se o item solicitado está no carrinho
            if item in carrinho:
                
                # Remove o item da lista carrinho
                carrinho.remove(item)
                
                # Exibe uma mensagem confirmando que o item foi removido
                print(f"'{item}' removido do carrinho.")
                
            else:
                
                # Informa ao usuário que o item não está no carrinho
                print(f"'{item}' não está no carrinho.")
                
        else:
            
            # Exibe uma mensagem informando que o carrinho está vazio
            print("O carrinho está vazio.")

    # Verifica se o comando digitado é 'ver'
    elif comando == 'ver':
        
        # Verifica se há itens no carrinho
        if carrinho:
            
            # Exibe os itens atualmente no carrinho
            print("Itens no carrinho:", carrinho)
            
        else:
            
            # Informa que o carrinho está vazio se não houver itens
            print("O carrinho está vazio.")

    # Verifica se o comando digitado é 'sair'
    elif comando == 'sair':
        
        # Exibe uma mensagem informando que o sistema está sendo encerrado
        print("Encerrando o sistema de carrinho.")
        
        # Sai do loop, terminando o programa
        break

    # Trata o caso de um comando que não seja nenhum dos anteriores
    else:
        
        # Informa ao usuário que o comando digitado é inválido
        print("Comando inválido. Tente novamente.")
```

```python
"""
25. Dada uma lista contendo os dias da semana na seguinte ordem:

['Terça', 'Quarta', 'Quinta', 'Sexta', 'Sabado', 'Domingo', 'Segunda']

Implemente um código em Python que reordene os elementos dessa lista de modo 
        que o primeiro elemento seja "Segunda", seguido pela sequência correta dos 
        dias da semana. O objetivo é reorganizar a lista para obter o seguinte resultado:

['Segunda', 'Terça', 'Quarta', 'Quinta', 'Sexta', 'Sabado', 'Domingo']
"""

# Lista original com os dias da semana, iniciando 
        # com 'Terça' e não em 'Segunda'
dias_da_semana = ['Terça', 'Quarta', 'Quinta', 'Sexta', 'Sabado', 'Domingo', 'Segunda']

# Utilizamos o método .index() para encontrar o 
        # índice (posição) onde está o dia 'Segunda' na lista
# A função index retorna a posição do primeiro elemento 
        # encontrado igual a 'Segunda'
indice_segunda = dias_da_semana.index('Segunda')

# Para reorganizar a lista, utilizamos o conceito 
        # de "slicing" (fatiamento) para dividir a lista em duas partes:
# A primeira parte é a sublista que começa em 'Segunda' 
        # até o final: dias_da_semana[indice_segunda:]
# A segunda parte é a sublista que vai do início até o índice 
        # da 'Segunda' (ou seja, os dias que vêm 
        # antes dela): dias_da_semana[:indice_segunda]
# Somando essas duas partes (concatenando as duas sublistas), 
        # conseguimos reorganizar a lista para começar em 'Segunda'
dias_reordenados = dias_da_semana[indice_segunda:] + dias_da_semana[:indice_segunda]

# Exibe a lista reorganizada com 'Segunda' como o 
        # primeiro dia, seguida dos demais dias em ordem correta
print("Lista reordenada:", dias_reordenados)
```

```python
"""
26. Exercício: "Jogo de Escape da Caverna"

Descrição: Você está preso em uma caverna com 3 portas misteriosas. 

Cada porta pode te levar a um destino diferente. O objetivo é encontrar a 
saída correta da caverna, mas existem armadilhas! O programa pedirá ao usuário 
para escolher uma porta a cada rodada e o levará a diferentes cenários. 

O jogo continua até que o usuário encontre a saída ou caia em uma armadilha.

Regras:

- Existem 3 portas: 1, 2, e 3.
- Apenas uma dessas portas leva à saída, as outras podem te levar 
            de volta ao início ou te fazer perder o jogo.
- O jogo dará opções de escolha ao usuário, e ele poderá decidir qual porta abrir.
- O jogador pode continuar explorando as portas até 
        # encontrar a saída ou ser pego por uma armadilha.
"""

import random  # Importa o módulo random para gerar escolhas aleatórias

# Função principal que controla o jogo
def jogar():
    
    # Exibe uma mensagem de boas-vindas ao jogador
    print("Bem-vindo ao Jogo de Escape da Caverna!")
    print("Você está preso em uma caverna escura com três portas misteriosas.\n")

    # Define aleatoriamente qual das 3 portas é a porta de saída
    # A função random.randint(1, 3) gera um número inteiro 
            # entre 1 e 3, que será a porta correta
    porta_saida = random.randint(1, 3)

    # Inicializa uma variável booleana que mantém o 
            # jogo em andamento
    # Enquanto jogando for True, o jogo continua
    jogando = True

    # Loop principal do jogo, que será repetido até o jogador 
            # encontrar a saída ou cair em uma armadilha
    while jogando:
        
        # Informa ao jogador para escolher uma porta
        print("Escolha uma porta para abrir (1, 2 ou 3):")
        
        # Solicita ao jogador que digite o número da porta que deseja abrir
        # O input é convertido para um número inteiro com int()
        escolha = int(input("Digite o número da porta: "))

        # Verifica se a escolha do jogador corresponde à porta de saída
        if escolha == porta_saida:
            
            # Se a porta escolhida for a porta de saída, o jogador vence o jogo
            print("Parabéns! Você encontrou a saída da caverna!")
            
            # O jogo termina, então a variável jogando é 
                    # definida como False para sair do loop
            jogando = False
            
        else:
            
            # Se o jogador não escolher a porta correta, ele será 
                    # levado a um destino alternativo
            # A função random.choice() seleciona aleatoriamente um 
                    # dos dois destinos possíveis
            destino = random.choice(["um túnel que te leva ao início", "uma armadilha perigosa! Você perdeu!"])
            
            # Verifica se o destino é uma armadilha
            if destino == "uma armadilha perigosa! Você perdeu!":
                
                # Se o jogador cair em uma armadilha, o jogo termina
                print("Você caiu em uma armadilha perigosa! O jogo terminou.")
                
                # O jogo termina, então jogando é definido como False
                jogando = False
                
            else:
                
                # Se o destino for um túnel, o jogador é enviado de volta ao início
                # O jogo continua e o jogador pode tentar novamente
                print(f"Você encontrou {destino}, tente novamente!\n")

# Inicia o jogo chamando a função jogar()
jogar()             
```

```python
"""
27. Exercício: "Quiz de Matemática com Desafios"

Descrição: Crie um programa em Python que faça perguntas de matemática ao usuário. 

O programa apresentará uma série de perguntas aleatórias, e o jogador terá 
        que responder corretamente para continuar. A cada resposta certa, o 
        nível de dificuldade aumenta, e o jogador ganha pontos. 
        
O jogo termina quando o jogador erra ou decide parar.

Regras:

- O jogo começa com operações simples (adição e subtração).
- Após cada resposta correta, o nível de dificuldade aumenta, 
        incluindo multiplicação e divisão.
- O jogador ganha pontos por cada resposta correta.
- O jogador pode parar o jogo a qualquer momento ou continuar 
        tentando aumentar sua pontuação.
- O jogo termina quando o jogador responde incorretamente ou escolhe sair.
"""

# Importa o módulo random para gerar números aleatórios.
import random

# Definição da função principal do quiz chamada 'quiz_matematica'.
def quiz_matematica():
    
    # Inicializa a variável 'pontuacao' para armazenar a 
            # pontuação do jogador, começando em 0.
    pontuacao = 0
    
    # Inicializa a variável 'nivel' para armazenar o nível 
            # atual do jogo, começando em 1.
    nivel = 1
    
    # Define uma variável booleana 'continuar_jogo' para 
            # controlar o loop principal do jogo.
    continuar_jogo = True

    # Exibe uma mensagem de boas-vindas ao usuário, 
            # indicando o início do quiz.
    print("Bem-vindo ao Quiz de Matemática!\n")
    
    # Inicia um loop que continuará enquanto a variável 
            # 'continuar_jogo' for True.
    while continuar_jogo:
        
        # Verifica se o jogador está no nível 1, que envolve 
                # operações de adição e subtração.
        if nivel == 1:
            
            # Gera um número inteiro aleatório 'num1' entre 1 e 10.
            num1 = random.randint(1, 10)
            
            # Gera outro número inteiro aleatório 'num2' entre 1 e 10.
            num2 = random.randint(1, 10)
            
            # Escolhe aleatoriamente uma operação entre 
                    # adição (+) e subtração (-).
            operacao = random.choice(["+", "-"])
            
            # Avalia a operação escolhida e calcula o resultado.
            if operacao == "+":
                
                # Se a operação for adição, soma 'num1' e 'num2'.
                resultado = num1 + num2
                
            else:
                
                # Se a operação for subtração, subtrai 'num2' de 'num1'.
                resultado = num1 - num2
                
        # Condição para níveis superiores, que envolve 
                # operações de multiplicação e divisão.
        else:
            
            # Gera um número aleatório entre 1 e 20 para o 
                    # primeiro número, num1.
            # Escolhe-se um intervalo maior para aumentar a 
                    # dificuldade conforme o nível aumenta.
            num1 = random.randint(1, 20)

            # Gera um número aleatório entre 1 e 10 para o 
                    # segundo número, num2.
            # Esse intervalo é mantido menor para simplificar as 
                    # operações, especialmente a divisão.
            num2 = random.randint(1, 10)

            # Escolhe aleatoriamente a operação entre multiplicação (*) e divisão (/).
            # Isso é feito para variar as operações e aumentar o desafio no jogo.
            operacao = random.choice(["*", "/"])

            # Calcula o resultado baseado na operação selecionada.
            if operacao == "*":
                
                # Se a operação escolhida for multiplicação, 
                        # multiplica num1 por num2.
                resultado = num1 * num2
                
            else:
                
                # Antes de dividir, ajusta num1 para ser o produto 
                        # de num1 e num2.
                # Isso garante que o resultado da divisão seja 
                        # sempre um número inteiro, evitando divisões 
                        # que resultem em frações ou números decimais.
                num1 = num1 * num2

                # Realiza a divisão inteira de num1 por num2.
                # A divisão inteira descarta o resto, proporcionando 
                        # um resultado limpo e fácil de adivinhar.
                resultado = num1 // num2

        # Exibe a pergunta formatada com a operação corrente para o usuário.
        print(f"Pergunta {nivel}: Quanto é {num1} {operacao} {num2}?")

        # Solicita a resposta do usuário, lê como string e 
                # converte imediatamente para inteiro.
        resposta = int(input("Resposta: "))

        # Verifica se a resposta fornecida pelo usuário é 
                # igual ao resultado calculado.
        if resposta == resultado:
            
            # Se correta, adiciona 10 pontos à pontuação do jogador.
            pontuacao += 10
            
            # Exibe uma mensagem de confirmação de resposta correta e 
                    # mostra a pontuação atual.
            print(f"Correto! Sua pontuação: {pontuacao} pontos.\n")
            
            # Aumenta o nível, indicando que o jogador passou 
                    # para uma fase mais difícil.
            nivel += 1
            
        else:
            
            # Se a resposta for incorreta, informa ao jogador a resposta correta.
            print(f"Resposta incorreta! A resposta correta era {resultado}.")
            
            # Encerra o jogo definindo 'continuar_jogo' como 
                    # False e saindo do loop.
            continuar_jogo = False
            break

        # Após cada rodada, pergunta ao usuário se deseja continuar o jogo.
        continuar = input("Deseja continuar? (s/n): ").lower()
        
        # Se o usuário decidir não continuar, ajusta a variável 
                # para encerrar o loop principal.
        if continuar != 's':
            continuar_jogo = False

    # Ao terminar o loop, exibe a pontuação final do jogador e 
            # uma mensagem de encerramento do jogo.
    print(f"\nJogo encerrado! Sua pontuação final: {pontuacao} pontos.")


# Chama a função para iniciar o jogo.
quiz_matematica()
```

```python
"""
28. Exercício: "Construa a Palavra"

Descrição: Neste jogo, o computador começa com uma letra válida, que 
        pertence a pelo menos uma das palavras no dicionário. 
        
O jogador deve continuar construindo a palavra com letras que ele 
        acredita que completam essa palavra válida. 
        
O jogo termina quando o jogador finaliza a palavra ou comete um erro, e 
        o programa verifica se a palavra construída é válida.

Regras:

- O jogo começa com uma letra válida que está no dicionário de palavras.
- O jogador deve adicionar letras para formar uma palavra.
- Se a letra escolhida não for possível para formar uma 
        palavra válida, o jogador será avisado.
- O jogador pode "finalizar" a palavra quando achar que ela está completa.
"""

# Importa o módulo random para permitir a seleção aleatória.
import random

# Lista contendo um conjunto de palavras válidas para o jogo.
dicionario_palavras = ["casa", "carro", "jogo", "animal", "caminho", "livro", "escola", "planta"]

# Função para escolher uma letra inicial de uma 
        # palavra aleatória do dicionário.
def escolher_letra_inicial(dicionario):
    
    # Escolhe uma palavra aleatoriamente do dicionário fornecido.
    palavra_escolhida = random.choice(dicionario)
    
    # Retorna o primeiro caractere da palavra escolhida, 
            # convertido para maiúscula.
    return palavra_escolhida[0].upper()

# Função principal do jogo que implementa a lógica 
        # de construção da palavra.
def construir_palavra():
    
    # Chama a função para obter uma letra inicial válida e 
            # armazena na variável 'letra_inicial'.
    letra_inicial = escolher_letra_inicial(dicionario_palavras)
    
    # Imprime uma mensagem de boas-vindas e mostra a 
            # letra inicial do jogo.
    print(f"Bem-vindo ao jogo 'Construa a Palavra'!")
    print(f"A primeira letra é: \"{letra_inicial}\"\n")

    # Inicializa 'palavra_atual' com a letra inicial já 
            # convertida para minúscula.
    palavra_atual = letra_inicial.lower()

    # Loop infinito para permitir ao usuário adicionar 
            # letras à palavra ou finalizar o jogo.
    while True:
        
        # Solicita ao usuário que digite uma letra ou a 
                # palavra 'fim' para terminar.
        # A função 'input' captura a entrada do usuário como uma string.
        # O método 'lower()' é usado para converter a entrada em 
                # letras minúsculas, garantindo consistência e 
                # facilitando a validação.
        letra = input("Digite uma letra para adicionar ou 'fim' para terminar a palavra: ").lower()
    
        # Condição para verificar se o usuário deseja terminar o jogo.
        if letra == "fim":
            
            # Se o usuário digitar 'fim', o loop 'while' é interrompido.
            # A instrução 'break' é usada para sair do loop, encerrando o 
                    # processo de adição de letras e movendo para a fase 
                    # de validação da palavra construída.
            break
            
        # Verifica se a entrada é uma única letra alfabética.
        # 'letra.isalpha()' verifica se a entrada contém apenas 
                # caracteres alfabéticos (letras).
        # 'len(letra) == 1' garante que a entrada seja 
                # exatamente um único caractere.
        elif letra.isalpha() and len(letra) == 1:
            
            # Se a entrada for uma única letra válida, esta é adicionada 
                    # ao final da string 'palavra_atual'.
            # O operador '+=' é utilizado para concatenar a nova letra à 
                    # sequência existente da palavra sendo formada.
            palavra_atual += letra
            
            # Exibe o estado atual da palavra sendo construída.
            # Utiliza-se 'f-string' para formatar a string de saída, 
                    # inserindo a 'palavra_atual' diretamente na 
                    # mensagem exibida ao usuário.
            print(f"A palavra atual é: \"{palavra_atual}\"\n")
            
        else:
            
            # Se a entrada for inválida (não é uma letra ou é mais de 
                    # um caractere e não é a palavra 'fim'),
                    # uma mensagem de erro é exibida, orientando o usuário a 
                    # inserir uma entrada válida.
            # Isso ajuda a evitar confusão ou entradas acidentais que 
                    # poderiam interromper a lógica do jogo.
            print("Entrada inválida. Digite uma letra ou 'fim'.")


    # Verifica se a palavra formada ao final do jogo está no
            # dicionário de palavras válidas.
    if palavra_atual in dicionario_palavras:
        
        # Mensagem de sucesso se a palavra existir no dicionário.
        print(f"Parabéns! \"{palavra_atual}\" é uma palavra válida! Você ganhou 10 pontos.")
        
    else:
        
        # Mensagem de erro se a palavra não for válida.
        print(f"\"{palavra_atual}\" não é uma palavra válida. Tente novamente!")

# Chama a função para iniciar o jogo.
construir_palavra()
```

```python
"""
29. Exercício: "Criptografia de Substituição"

Descrição: Neste exercício, o jogador deve decifrar uma mensagem criptografada 
        usando uma substituição de letras. O programa gera uma mensagem oculta, 
        onde cada letra foi substituída por outra letra. 
        
O jogador deve tentar adivinhar qual é a substituição correta, letra por letra, 
        para revelar a mensagem original. A cada tentativa, o programa revela as 
        letras corretas ou informa se o jogador errou.

Regras:

- O jogador recebe uma mensagem criptografada com substituição de 
        letras (exemplo: 'A' foi substituído por 'X', 'B' por 'Y', etc.).
- O jogador deve tentar adivinhar a correspondência entre as letras.
- O programa verifica a resposta e atualiza a mensagem com as 
        letras corretas a cada tentativa.
- O jogo termina quando a mensagem inteira for decifrada.
"""

# Importa os módulos necessários para o jogo.
import random
import string

# Função para criar um mapa de substituição de letras aleatórias.
def gerar_criptografia():
    
    """
    Esta função gera um dicionário de substituição de letras, 
            onde cada letra do alfabeto é mapeada para outra letra aleatória, 
            sem repetições, efetivamente criando uma cifra de 
            substituição monoalfabética.
    """

    # Gera uma lista das letras do alfabeto em minúscula.
    # 'string.ascii_lowercase' é uma string pré-definida em 
            # Python que contém todas as letras de 'a' a 'z'.
    # Ao converter essa string para uma lista, cada letra se 
            # torna um elemento separado da lista.
    letras = list(string.ascii_lowercase)
    
    # Embaralha a lista de letras para criar uma sequência aleatória.
    # 'random.shuffle' recebe uma lista e reorganiza seus 
            # elementos em ordem aleatória.
    # Isso é crucial para garantir que a cifra de 
            # substituição não seja previsível.
    random.shuffle(letras)
    
    # Cria um dicionário que mapeia cada letra do alfabeto a 
            # uma letra embaralhada.
    # Este dicionário é criado usando compreensão de dicionário, 
            # que é uma forma concisa de construir dicionários em Python.
    # 'chr(i + 97)' converte um inteiro em um caractere, 
            # onde 97 é o código ASCII para 'a'.
    # O loop 'for i in range(26)' itera 26 vezes, 
            # correspondendo às 26 letras do alfabeto.
    # Para cada iteração, uma letra do alfabeto (a partir de 'a') é 
            # associada a uma letra na lista embaralhada.
    substituicao = {chr(i + 97): letras[i] for i in range(26)}
    
    # Retorna o dicionário de substituição.
    # Este dicionário será usado para substituir as letras na 
            # mensagem original durante a criptografia.
    return substituicao


# Função para criptografar uma mensagem usando o mapa de substituição.
def criptografar_mensagem(mensagem, substituicao):
    
    """
    Esta função aceita uma 'mensagem' (string) e um 'substituicao' (dicionário) 
            que contém o mapa de substituição de letras.
    Retorna a mensagem criptografada onde cada letra é substituída 
            por sua correspondente no dicionário.
    """

    # A função utiliza uma compreensão de lista para iterar sobre 
            # cada caractere 'c' na 'mensagem'.
    # O método 'substituicao.get(c, c)' tenta obter a substituição 
            # para o caractere 'c'.
    # Se 'c' é uma chave no dicionário 'substituicao', retorna o 
            # valor correspondente (letra substituta).
    # Se 'c' não é uma chave no dicionário (como no caso de espaços, 
            # pontuação ou caracteres que não foram mapeados),
    # retorna 'c' como está, mantendo o caractere original na 
            # mensagem criptografada.
    # 'join' é usado para concatenar todos os caracteres 
            # processados de volta em uma única string.
    return ''.join(substituicao.get(c, c) for c in mensagem)


# Função para decifrar a mensagem com as letras adivinhadas pelo usuário.
def decifrar_mensagem(mensagem, mapeamento_usuario):
    
    """
    Esta função decifra a mensagem criptografada usando as 
            adivinhações do usuário armazenadas em 'mapeamento_usuario'.
    Cada letra criptografada é substituída pela letra adivinhada 
            correspondente ou, se não adivinhada, por '*'.
    """

    # A função também usa uma compreensão de lista para processar 
            # cada caractere 'c' na 'mensagem'.
    # 'mapeamento_usuario.get(c, '*')' tenta obter a letra 
            # adivinhada para o caractere 'c'.
    # Se 'c' é uma letra e está no 'mapeamento_usuario', 
            # retorna a letra adivinhada.
    # Se 'c' não foi adivinhado ou não é uma letra (como espaços 
            # ou pontuação), verifica-se com 'c.isalpha()'.
    # Se 'c.isalpha()' for False, o caractere 'c' é retornado como está.
    # Caso contrário, se não houver adivinhação para 'c', retorna '*' 
            # para indicar que a letra ainda precisa ser adivinhada.
    return ''.join(mapeamento_usuario.get(c, '*') if c.isalpha() else c for c in mensagem)


# Função principal que organiza a lógica do jogo de criptografia.
def jogo_criptografia():
    
    """
    Esta função coordena o fluxo do jogo de criptografia, 
            onde uma mensagem é criptografada
            e o jogador deve adivinhar a substituição 
            de letras para decifrá-la.
    """

    # Define a mensagem original que será criptografada.
    # Neste exemplo, a mensagem é uma string simples 
            # para facilitar o entendimento.
    mensagem_original = "este e um teste"
    
    # Gera o mapa de substituição de letras utilizando a 
            # função definida anteriormente.
    # Esta função cria um mapeamento aleatório de cada letra 
            # do alfabeto para outra letra.
    substituicao = gerar_criptografia()
    
    # Criptografa a mensagem original usando o mapa de 
            # substituição gerado.
    # A mensagem resultante será uma versão da mensagem original 
            # onde cada letra foi substituída
            # conforme o mapeamento definido em 'substituicao'.
    mensagem_criptografada = criptografar_mensagem(mensagem_original, substituicao)

    # Exibe a mensagem criptografada para o jogador, dando a 
            # ele o ponto de partida para decifrar.
    print(f"Mensagem criptografada: {mensagem_criptografada}")
    
    # Inicializa um dicionário para armazenar as adivinhações do 
            # usuário sobre a substituição de letras.
    mapeamento_usuario = {}

    # Decifra inicialmente a mensagem criptografada 
            # substituindo letras conhecidas pelo usuário
    # e marcando letras desconhecidas com '*' para indicar 
            # que ainda não foram decifradas.
    mensagem_decifrada = decifrar_mensagem(mensagem_criptografada, mapeamento_usuario)
    print(f"Mensagem decifrada até agora: {mensagem_decifrada}\n")

    # Loop de adivinhação onde o usuário tenta identificar 
            # as substituições corretas.
    while True:
        
        # Solicita ao usuário que digite a letra criptografada 
                # que deseja substituir.
        letra_criptografada = input("Digite a letra criptografada que você quer decifrar: ").lower()
        
        # Solicita ao usuário a letra que ele acredita ser a 
                # original correspondente à criptografada.
        letra_real = input(f"Qual é a letra original correspondente a '{letra_criptografada}'?: ").lower()

        # Adiciona a correspondência entre a letra criptografada e a 
                # suposta letra real ao mapeamento do usuário.
        mapeamento_usuario[letra_criptografada] = letra_real

        # Atualiza a mensagem decifrada com as novas 
                # adivinhações do usuário.
        mensagem_decifrada = decifrar_mensagem(mensagem_criptografada, mapeamento_usuario)
        print(f"\nMensagem decifrada até agora: {mensagem_decifrada}\n")

        # Verifica se a mensagem foi completamente decifrada, ou 
                # seja, se não contém mais o caractere '*'.
        if '*' not in mensagem_decifrada:
            
            # Se a mensagem está completamente decifrada, exibe uma 
                    # mensagem de parabéns e termina o loop.
            print("Parabéns! Você decifrou a mensagem completa!")
            break

# Chamada da função para iniciar o jogo.
jogo_criptografia()
```

```python
"""
30. Exercício: "Calculadora de Tabuada Interativa"

Descrição: Neste jogo, o jogador escolhe um número e a calculadora gera a 
        tabuada daquele número. Em vez de simplesmente exibir a tabuada completa, o 
        jogo faz perguntas e o jogador deve fornecer as respostas corretas para 
        cada multiplicação. O objetivo é acertar todas as respostas.

Regras:

- O jogador escolhe um número para gerar a tabuada.
- O programa faz perguntas de multiplicação para cada valor de 1 a 10.
- O jogador deve responder corretamente a cada pergunta.
- No final, o jogador recebe uma pontuação com base no número de respostas corretas.
"""

# Função principal do jogo de tabuada interativa.
def tabuada_interativa():
    
    """
    Esta função conduz um jogo onde o jogador escolhe 
            um número e responde a perguntas
            de multiplicação relacionadas à tabuada 
            desse número, de 1 a 10. O jogo avalia as
            respostas e fornece feedback.
    """

    # Mensagem de boas-vindas ao jogador.
    print("Bem-vindo à Calculadora de Tabuada Interativa!")
    
    # Solicita ao jogador que escolha um número para 
            # praticar a tabuada.
    # O input captura a entrada como string e a 
            # função int() converte para um número inteiro.
    numero = int(input("Digite o número para o qual você deseja praticar a tabuada: "))
    
    # Inicializa um contador para o número de acertos.
    acertos = 0
    
    # Loop que itera sobre os números de 1 a 10 para 
            # criar perguntas de tabuada.
    for i in range(1, 11):
        
        # Calcula a resposta correta para a pergunta atual.
        resposta_correta = numero * i
        
        # Solicita ao jogador que forneça uma resposta 
                # para a multiplicação atual.
        resposta_usuario = int(input(f"Quanto é {numero} x {i}? "))
        
        # Verifica se a resposta fornecida pelo 
                # jogador está correta.
        if resposta_usuario == resposta_correta:
            
            # Se correta, imprime uma mensagem de confirmação e 
                    # incrementa o contador de acertos.
            print("Correto!\n")
            acertos += 1
            
        else:
            
            # Se incorreta, informa ao jogador a resposta correta.
            print(f"Incorreto. A resposta certa é {resposta_correta}.\n")
    
    # Após o loop, exibe o resultado final, informando 
            # quantas respostas foram corretas.
    print(f"Parabéns! Você acertou {acertos} de 10 respostas!")

# Chamada da função para iniciar o jogo.
tabuada_interativa()
```

```python
"""
32. Exercício: "Jogo da Memória Numérica"

Descrição: Neste jogo, o jogador deve memorizar uma sequência de números 
        que será exibida por alguns segundos. Após o tempo, a sequência 
        desaparece e o jogador deve tentar reproduzir a sequência corretamente. 
        
O jogo aumenta a dificuldade a cada rodada, adicionando um 
        número a mais na sequência.

Regras:

- O jogo começa com uma sequência de 3 números.
- O jogador tem 3 segundos para memorizar a sequência antes que ela desapareça.
- A cada rodada, a sequência aumenta em um número, e o jogador 
            deve tentar lembrá-la corretamente.
- O jogo continua até o jogador errar a sequência.
"""

# Importa os módulos necessários para o jogo.
import random
import time

# Função principal que organiza a lógica do jogo 
        # de memória numérica.
def jogo_memoria_numerica():
    
    """
    Esta função controla o fluxo do jogo da memória numérica, 
            onde o jogador deve memorizar e reproduzir uma sequência 
            de números que se torna progressivamente mais longa.
    """

    # Mensagem de boas-vindas ao jogador.
    print("Bem-vindo ao Jogo da Memória Numérica!")
    
    # Inicializa a sequência começando com três números.
    tamanho_sequencia = 3
    
    # Loop principal que continua até o jogador errar a sequência.
    while True:
        
        # Gera uma nova sequência de números aleatórios com o 
                # tamanho atual da sequência.
        # A compreensão de lista cria uma lista de inteiros 
                # aleatórios entre 0 e 9.
        sequencia = [random.randint(0, 9) for _ in range(tamanho_sequencia)]
        
        # Mostra a sequência para o jogador e pede para memorizá-la.
        # A função 'join' é usada para converter os números em 
                # strings e concatená-los com espaços.
        print(f"\nMemorize esta sequência: {' '.join(map(str, sequencia))}")
        
        # Pausa a execução por 3 segundos, dando tempo para o 
                # jogador memorizar a sequência.
        time.sleep(3)
        
        # Limpa a tela para simular o desaparecimento da sequência.
        # Este comando é específico para terminais que suportam 
                # códigos ANSI, como o terminal padrão no Linux ou MacOS.
        print("\033[H\033[J")

        # Solicita ao jogador que digite a sequência memorizada.
        resposta = input("Digite a sequência: ").split()

        # Verifica se a resposta do jogador está correta.
        # A função 'map(int, resposta)' converte as 
                # entradas de string para inteiros.
        # Compara a lista de respostas convertida com a 
                # sequência original.
        if list(map(int, resposta)) == sequencia:
            
            print("Correto!\n")
            
            # Se correta, aumenta o tamanho da sequência 
                    # para a próxima rodada.
            tamanho_sequencia += 1
            
        else:
            
            # Se incorreta, informa ao jogador a sequência 
                    # correta e termina o jogo.
            print(f"Incorreto! A sequência correta era {' '.join(map(str, sequencia))}.")
            print("Fim de jogo!")
            break

# Chamada da função para iniciar o jogo.
jogo_memoria_numerica()
```

```python
"""
32. Exercício: "Jogo de Criação de Palavras"

Descrição: Neste jogo, o jogador recebe várias letras aleatórias e precisa 
        formar a maior quantidade de palavras possível a partir dessas letras. 
        
O jogador tem um limite de tempo para formar as palavras e ganha pontos 
        por cada palavra válida. O jogo termina quando o tempo acaba, e a 
        pontuação final é calculada com base no número de palavras formadas.

Regras:

- O jogador recebe 4 letras aleatórias.
- O jogador tem 60 segundos para formar quantas palavras 
            puder com essas letras.
- Cada palavra formada corretamente dá pontos ao jogador.
"""

import random
import time

# Importa o módulo threading para utilizar temporizadores.
import threading  

# Dicionário simples de palavras válidas usado no jogo.
dicionario_palavras = ["rato", "esto", "ator", "tora", "rosa", "arte", "teor", "rota", "ser", "seta"]

# Função para gerar um conjunto de letras aleatórias a 
        # partir das palavras disponíveis no dicionário.
def gerar_letras_validas(dicionario):
    
    """
    Esta função gera um conjunto de letras únicas que podem ser 
            usadas para formar palavras válidas.
    Ela escolhe uma palavra aleatoriamente do dicionário 
            fornecido e extrai letras únicas dessa palavra,
            garantindo que o jogador tenha a chance de formar 
            pelo menos uma palavra válida completa.
    """

    todas_letras = []  # Lista para armazenar todas as letras das palavras escolhidas.
    
    # Seleciona uma palavra aleatória do dicionário.
    # 'random.choice' é usado para escolher um item 
            # aleatório de uma lista.
    palavra = random.choice(dicionario)
    
    # Adiciona cada letra da palavra escolhida à lista 'todas_letras'.
    # 'list(palavra)' converte a string da palavra em 
            # uma lista de caracteres.
    todas_letras.extend(list(palavra))
    
    # Cria uma lista de letras únicas a partir de 'todas_letras'.
    # 'set(todas_letras)' converte a lista em um conjunto, 
            # que automaticamente remove quaisquer duplicatas.
    letras_unicas = list(set(todas_letras))
    
    # Retorna uma amostra das letras únicas.
    # 'random.sample' é usado para selecionar uma quantidade 
            # específica de itens de uma lista sem repetição.
    # 'min(6, len(letras_unicas))' garante que o número de letras 
            # escolhidas não exceda o número de letras únicas 
            # disponíveis ou seis.
    return random.sample(letras_unicas, min(6, len(letras_unicas)))


# Função para verificar se uma palavra pode ser formada 
        # com as letras disponíveis.
def palavra_valida(palavra, letras_disponiveis):
    
    """
    Esta função determina se a palavra fornecida pelo jogador 
            pode ser formada usando as letras disponíveis.
    Ela verifica cada letra na palavra para garantir que a quantidade 
            dessa letra na palavra não exceda
            a quantidade dessa mesma letra nas letras disponíveis.
    """

    # A função 'all()' verifica se todos os elementos de um 
            # iterável são verdadeiros.
    # A expressão interna 'palavra.count(letra) <= letras_disponiveis.count(letra)' é uma comparação
            # que conta quantas vezes uma letra ocorre na 'palavra' e 
            # compara isso com o número de vezes
            # que a mesma letra ocorre nas 'letras_disponiveis'.
    # O loop 'for letra in palavra' percorre cada letra na 
            # palavra proposta pelo jogador.
    # Se todas as comparações são verdadeiras (i.e., se a 
            # palavra pode ser formada com as letras disponíveis),
            # a função retorna True. Se qualquer letra na 
            # palavra excede a quantidade disponível, retorna False.
    return all(palavra.count(letra) <= letras_disponiveis.count(letra) for letra in palavra)


# Função para encerrar o jogo após o tempo limite.
def encerrar_jogo():
    
    """
    Esta função é chamada quando o temporizador do jogo expira.
    Ela define a variável global 'jogo_ativo' como False, o que é 
            verificado no loop principal do jogo
            para encerrar a sessão de jogo.
    """

    # Declara que a função modificará a variável global 'jogo_ativo'.
    global jogo_ativo  

    # Altera o valor de 'jogo_ativo' para False, indicando 
            # que o jogo terminou.
    jogo_ativo = False  




# Definição da função principal do jogo.
def jogo_criacao_palavras():
    
    """
    Gerencia o jogo de criação de palavras onde o jogador forma 
            palavras a partir de letras aleatórias.
    O jogador tem um tempo limitado para formar as palavras e 
            ganha pontos por cada palavra válida formada.
    """

    # Chama a função para gerar um conjunto de letras aleatórias que 
            # podem ser usadas para formar palavras.
    letras_disponiveis = gerar_letras_validas(dicionario_palavras)
    
    # Imprime uma mensagem de boas-vindas e as regras básicas do jogo.
    print("Bem-vindo ao Jogo de Criação de Palavras!")
    print("Você tem 60 segundos para formar palavras.")
    print(f"Letras disponíveis: {', '.join(letras_disponiveis).upper()}\n")

    # Define uma variável global 'jogo_ativo' para controlar o 
            # fluxo do jogo dentro do loop.
    global jogo_ativo

    # Inicializa o jogo como ativo.
    jogo_ativo = True  
    
    # Inicializa variáveis para rastrear a pontuação e as 
            # palavras que o jogador já formou.
    pontuacao = 0
    palavras_formadas = []

    # Configura um temporizador de 60 segundos que chamará a 
            # função 'encerrar_jogo' quando o tempo expirar.
    temporizador = threading.Timer(10, encerrar_jogo)

    # Ativa o temporizador para iniciar a contagem regressiva.
    temporizador.start()  

    # Loop principal do jogo que continua enquanto 'jogo_ativo' for True.
    while jogo_ativo:
        
        # Solicita ao jogador para inserir uma palavra 
                # usando as letras disponíveis.
        palavra = input("Digite uma palavra: ").lower()

        # Verifica se o temporizador já terminou para parar 
                # de aceitar novas entradas.
        if not jogo_ativo:
            break

        # Checa se a palavra inserida está no dicionário de palavras 
                # válidas e se pode ser formada com as letras disponíveis.
        if palavra in dicionario_palavras and palavra_valida(palavra, letras_disponiveis):
            
            # Confere se a palavra já foi formada anteriormente 
                    # para evitar duplicatas.
            if palavra not in palavras_formadas:

                # Adiciona a palavra válida à lista de palavras formadas.
                palavras_formadas.append(palavra)  
                
                # Adiciona pontos à pontuação do jogador.
                pontuacao += 10  
                print(f"Correto! +10 pontos.\n")
                
            else:
                
                # Informa ao jogador que a palavra já foi usada.
                print("Você já formou essa palavra.\n")
                
        else:
            
            # Informa ao jogador que a palavra é inválida.
            print("Incorreto! Essa palavra não pode ser formada com as letras disponíveis.\n")

    # Após o término do jogo, mostra as palavras possíveis 
            # que poderiam ter sido formadas.
    # Esta seção é executada depois que o temporizador 
            # termina e o loop do jogo é encerrado.
    palavras_possiveis = [

        # Itera sobre cada palavra no dicionário de 
                # palavras válido.
        palavra for palavra in dicionario_palavras  

        # Inclui a palavra na lista se ela pode ser formada 
                # com as letras disponíveis.
        if palavra_valida(palavra, letras_disponiveis)  
        
    ]
    
    # A lista 'palavras_possiveis' agora contém todas as palavras do 
            # dicionário que o jogador poderia ter formado
    # com as letras que foram dadas, considerando que elas atendem às 
            # regras de formação de palavras estabelecidas
            # pela função 'palavra_valida'. Esta função verifica se a 
            # quantidade de cada letra na palavra não excede
            # a quantidade dessa letra nas letras disponíveis.
    
    # Exibe o resultado final, incluindo o número total de 
            # palavras formadas e a pontuação.
    # A mensagem final fornece feedback ao jogador sobre seu 
            # desempenho e também oferece uma visão do que poderia
            # ter sido alcançado, o que pode ser motivador e educativo.
    print(f"\nVocê formou {len(palavras_formadas)} palavras e ganhou {pontuacao} pontos.")

    # Mostra a lista de palavras que o jogador poderia ter formado. 
    # Isso pode ajudar o jogador a aprender novas palavras
            # e a entender melhor como usar as letras de maneira eficaz.
    print(f"Palavras que poderiam ser formadas com as letras disponíveis: {', '.join(palavras_possiveis)}")


# ["rato", "esto", "ator", "tora", "rosa", "arte", "teor", "rota", "ser", "seta"]

# Inicia o jogo
jogo_criacao_palavras()
```
