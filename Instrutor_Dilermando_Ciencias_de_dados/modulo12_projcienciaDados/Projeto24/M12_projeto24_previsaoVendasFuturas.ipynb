## Modelo de Previsão de Vendas Futuras

Vamos treinar o modelo utilizando o algoritmo de regressão linear. Antes, vamos dividir nosso dataset em duas partes: x e y, ou seja, x com os atributos ou características do conjunto de dados e y com a variável que representa as saídas desejadas.  Depois disso, dividimos em mais duas partes: treino e teste.

```python
y = np.array(data["Sales"])
x = np.array(data.drop(['Sales'], axis=1))
```

```python
# agora dividindo em treino e teste
xtrain, xtest, ytrain, ytest = train_test_split(x, y, test_size=0.2, random_state=42)
```

Vamos agora treinar o modelo usando o algoritmo de Regressão Linear

```python
model = LinearRegression()
model.fit(xtrain, ytrain)
print(model.score(xtest, ytest))
```

Com uma acurácia de **90%**, o nosso modelo está pronto para fazer previsões...

```python
#features = [[TV, Radio, Newspaper]]
features = np.array([[230.1, 37.8, 69.2]])
print(model.predict(features))
```
