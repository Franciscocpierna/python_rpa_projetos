## 2. Funções

Uma função é uma relação matemática que associa cada elemento de um conjunto (domínio) a um único elemento de outro conjunto (contradomínio). Em outras palavras, uma função é uma regra que atribui valores de saída a valores de entrada.

**2.1 O que é uma função?**

Uma função é uma relação matemática que associa cada elemento de um conjunto (domínio) a um único elemento de outro conjunto (contradomínio). Em outras palavras, uma função é uma regra que atribui valores de saída a valores de entrada.

Uma função é representada por uma expressão matemática que envolve uma ou mais variáveis, e o resultado dessa expressão é o valor de saída da função. O conjunto de todos os valores possíveis de entrada é chamado de domínio, e o conjunto de todos os valores possíveis de saída é chamado de contradomínio.

Em termos visuais, podemos imaginar uma função como um "mecanismo" que transforma um valor de entrada em um valor de saída, seguindo uma determinada regra ou processo.

As funções são amplamente utilizadas em matemática, ciências, engenharia e em muitos outros campos, pois fornecem uma maneira de descrever e analisar relações entre variáveis e fazer previsões com base em dados.

**2.2 Aplicação de uma função - Estudo do Vinho**

Vamos utilizar o exemplo do vinho para ilustrar a aplicação de uma função em diversas óticas, como química, saúde e amizade. Vamos considerar que queremos analisar a relação entre a concentração de álcool no vinho e seu envelhecimento, e como esses fatores podem afetar diferentes aspectos.

Suponhamos que temos uma função matemática que descreve a concentração de álcool (em porcentagem) no vinho em relação ao tempo de envelhecimento (em anos):

#### Concentracao_de_alcool(envelhecimento) = 14 - 0.2 * envelhecimento

Nessa função, envelhecimento é a variável de entrada (domínio) que representa o tempo em anos, e Concentracao_de_alcool(envelhecimento) é o valor de saída (contradomínio) que representa a concentração de álcool em porcentagem.

Agora, vamos explorar algumas análises utilizando essa função em Python...

```python
import matplotlib.pyplot as plt
import numpy as np

# Função que descreve a concentração de álcool no vinho em relação ao envelhecimento
def Concentracao_de_alcool(envelhecimento):
    return 14 - 0.2 * envelhecimento

# Valores para o envelhecimento do vinho (de 0 a 50 anos)
envelhecimento = np.linspace(0, 50, 100)

# Calculando os valores de concentração de álcool usando a função
concentracao = Concentracao_de_alcool(envelhecimento)

# Plotando o gráfico da concentração de álcool versus envelhecimento
plt.plot(envelhecimento, concentracao)
plt.xlabel('Envelhecimento (anos)')
plt.ylabel('Concentração de Álcool (%)')
plt.title('Concentração de Álcool no Vinho com o Envelhecimento')
plt.grid(True)
plt.show()
```

Agora, vamos explorar outras análises do vinho sob diferentes óticas:

**Análise Farmacológica:**

Digamos que temos outra função que descreve como a concentração de álcool no sangue de uma pessoa diminui ao longo do tempo após a ingestão de uma quantidade específica de vinho:

#### Concentracao_no_sangue(tempo) = 0.1 * tempo

Nessa função, tempo é a variável de entrada (domínio) que representa o tempo em horas após a ingestão, e Concentracao_no_sangue(tempo) é o valor de saída (contradomínio) que representa a concentração de álcool no sangue.

Vamos plotar o gráfico dessa função em Python:

```python
# Função que descreve a concentração de álcool no sangue em relação ao tempo após ingestão
def Concentracao_no_sangue(tempo):
    return 0.1 * tempo

# Valores para o tempo após ingestão do vinho (de 0 a 5 horas)
tempo = np.linspace(0, 5, 100)

# Calculando os valores de concentração de álcool no sangue usando a função
concentracao_sangue = Concentracao_no_sangue(tempo)

# Plotando o gráfico da concentração de álcool no sangue versus tempo após ingestão
plt.plot(tempo, concentracao_sangue)
plt.xlabel('Tempo após ingestão (horas)')
plt.ylabel('Concentração de Álcool no Sangue (%)')
plt.title('Concentração de Álcool no Sangue após Ingestão de Vinho')
plt.grid(True)
plt.show()
```

**Análise Social - Amizade e Vinho:**

Vamos considerar uma função que descreve o nível de amizade (de 0 a 100) entre duas pessoas em relação à quantidade de vinho compartilhado em um período de tempo:

#### Nivel_de_amizade(qtd_vinho_compartilhado) = 50 + 2 * qtd_vinho_compartilhado

Nessa função, qtd_vinho_compartilhado é a variável de entrada (domínio) que representa a quantidade de vinho compartilhado em litros, e Nivel_de_amizade(qtd_vinho_compartilhado) é o valor de saída (contradomínio) que representa o nível de amizade entre as pessoas.

Vamos plotar o gráfico dessa função em Python:

```python
# Função que descreve o nível de amizade em relação à quantidade de vinho compartilhado
def Nivel_de_amizade(qtd_vinho_compartilhado):
    return 50 + 2 * qtd_vinho_compartilhado

# Valores para a quantidade de vinho compartilhado (de 0 a 10 litros)
qtd_vinho_compartilhado = np.linspace(0, 10, 100)

# Calculando os valores de nível de amizade usando a função
nivel_amizade = Nivel_de_amizade(qtd_vinho_compartilhado)

# Plotando o gráfico do nível de amizade versus quantidade de vinho compartilhado
plt.plot(qtd_vinho_compartilhado, nivel_amizade)
plt.xlabel('Quantidade de Vinho Compartilhado (litros)')
plt.ylabel('Nível de Amizade')
plt.title('Nível de Amizade em Relação à Quantidade de Vinho Compartilhado')
plt.grid(True)
plt.show()
```

Esses exemplos ilustram como as funções podem ser aplicadas para analisar e entender o vinho sob diversas óticas, como química, saúde e amizade. 

Em cada análise, temos uma função diferente que descreve a relação entre as variáveis de interesse. Os gráficos ajudam a visualizar essas relações e proporcionam uma melhor compreensão dos dados.

**2.3 Funções clássicas**

Existem várias funções matemáticas clássicas que são frequentemente encontradas em problemas e aplicações. Algumas das funções clássicas incluem:

•	**Função Linear**: f(x) = mx + b, onde m é a inclinação (coeficiente angular) e b é o intercepto no eixo y.

•	**Função Quadrática**: f(x) = ax^2 + bx + c, onde a, b e c são constantes e a ≠ 0.

•	**Função Exponencial**: f(x) = a^x, onde a é uma constante positiva.

•	**Função Logarítmica**: f(x) = log_a(x), onde a é a base do logaritmo e x > 0.

*Definição e exemplos de Funções em Python*

Em Python, você pode definir funções utilizando a sintaxe *def*. Uma função em Python recebe um conjunto de argumentos (valores de entrada) e pode retornar um valor (valor de saída). Utilizaremos a biblioteca matplotlib para plotar os gráficos. 

**a) Exemplo de Função Linear**

Vamos criar uma função linear em Python que recebe o valor de x, a inclinação m e o intercepto b, e retorna o valor de y de acordo com a fórmula 

*f(x) = mx + b*.

```python
# Definindo uma função
def funcao_linear(x, m, b):
    y = m * x + b
    return y
```

Agora, podemos usar essa função para calcular valores de saída y com base nos valores de entrada x, m e b.

```python
# Exemplo de uso da função linear
x = 2
m = 3
b = 1

resultado = funcao_linear(x, m, b)
print(f"Para x = {x}, f(x) = {resultado}")
```

Plotando a Função Linear:

```python
import matplotlib.pyplot as plt

def funcao_linear(x, m, b):
    return m * x + b

# Valores para a função linear
m = 2
b = 5
x = range(-10, 11)  # Valores de x de -10 a 10

# Calculando os valores de y
y = [funcao_linear(i, m, b) for i in x]

# Plotando o gráfico
plt.plot(x, y)
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Gráfico da Função Linear')
plt.grid(True)
plt.show()
```

**b) Exemplo de Função Quadrática**

Vamos criar uma função em Python que recebe o valor de x, os coeficientes a, b e c, e retorna o valor de y de acordo com a fórmula 

*f(x) = ax^2 + bx + c.*


```python
# definindo uma função quadrática
def funcao_quadratica(x, a, b, c):
    y = a * x**2 + b * x + c
    return y
```

Agora, podemos usar essa função para calcular valores de saída y com base nos valores de entrada x, a, b e c.

```python
# Exemplo de uso da função quadrática
x = 3
a = 2
b = -3
c = 1

resultado = funcao_quadratica(x, a, b, c)
print(f"Para x = {x}, f(x) = {resultado}")
```

Plotando a Função Quadrática:

```python
import matplotlib.pyplot as plt

def funcao_quadratica(x, a, b, c):
    return a * x**2 + b * x + c

# Valores para a função quadrática
a = 1
b = -2
c = 1
x = range(-10, 11)  # Valores de x de -10 a 10

# Calculando os valores de y
y = [funcao_quadratica(i, a, b, c) for i in x]

# Plotando o gráfico
plt.plot(x, y)
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Gráfico da Função Quadrática')
plt.grid(True)
plt.show()
```

**c) Exemplo de Função Exponencial**

Em Python, a função exponencial pode ser facilmente implementada usando o operador **.

```python
# definindo uma função exponencial
def funcao_exponencial(x, a):
    y = a ** x
    return y
```

Agora, podemos usar essa função para calcular valores de saída y com base nos valores de entrada x e a.

```python
# Exemplo de uso da função exponencial
x = 2
a = 3

resultado = funcao_exponencial(x, a)
print(f"Para x = {x}, f(x) = {resultado}")
```

Plotando a Função Exponencial:

```python
import matplotlib.pyplot as plt

def funcao_exponencial(x, a):
    return a ** x

# Valores para a função exponencial
a = 2
x = range(-10, 11)  # Valores de x de -10 a 10

# Calculando os valores de y
y = [funcao_exponencial(i, a) for i in x]

# Plotando o gráfico
plt.plot(x, y)
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Gráfico da Função Exponencial')
plt.grid(True)
plt.show()
```

**d) Exemplo de Função Logarítmica**

Em Python, a função logarítmica pode ser implementada utilizando a biblioteca *math* que já possui a função *log()*.


```python
#definindo uma função logaritmica (utilizando uma já existente no pacote/biblioteca math)
import math

def funcao_logaritmica(x, a):
    y = math.log(x, a)
    return y
```

Agora, podemos usar essa função para calcular valores de saída y com base nos valores de entrada x e a.

```python
# Exemplo de uso da função logarítmica
x = 10
a = 2

resultado = funcao_logaritmica(x, a)
print(f"Para x = {x}, f(x) = {resultado}")
```

Plotando a Função Logarítmica:

```python
import matplotlib.pyplot as plt
import numpy as np

def funcao_logaritmica(x, a):
    return np.log(x) / np.log(a)

# Valores para a função logarítmica
a = 2
x = np.linspace(0.1, 10, 100)  # Valores de x de 0.1 a 10 (o log não é definido para x=0)

# Calculando os valores de y
y = [funcao_logaritmica(i, a) for i in x]

# Plotando o gráfico
plt.plot(x, y)
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Gráfico da Função Logarítmica')
plt.grid(True)
plt.show()
```
